```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Load and prepare image
data = np.load("C:\\Users\\Mesterpc.ir\\Downloads\\masks.npy")
img = cv2.cvtColor(data[1, :, :, :], cv2.COLOR_BGR2RGB)

# Parameters
MASK_SIZE = (16, 16)  # 16x16 pixels as specified
windows = {
    'top': (105, 21, 27, 39),    # (x, y, width, height)
    'left': (89, 126, 46, 57),   # Septal mitral annulus
    'right': (130, 119, 46, 57)  # Lateral mitral annulus
}

# Create anatomically accurate masks
def create_anatomical_masks():
    masks = {}
    size = 16  # 16x16 masks
    
    # Cardiac apex mask (sharp V-shape pointing down)
    M_top = np.zeros((size, size), dtype=np.float32)
    for i in range(size):
        width = max(1, int(2 + 6 * (i/(size-1))))  # Sharper V-shape
        start = max(0, (size//2) - width//2)
        end = min(size, (size//2) + width//2 + 1)
        M_top[i, start:end] = 1.0
    # Add focal point at apex tip (bottom center)
    M_top[12:16, 6:10] = 3.0
    masks['top'] = M_top

    # Septal mitral annulus (right-biased L-shape)
    M_septal = np.zeros((size, size), dtype=np.float32)
    # Vertical bar (septal wall)
    M_septal[8:14, 10:14] = 1.0
    # Horizontal bar (valve plane)
    M_septal[12:14, 6:14] = 1.0
    # Focal point at annulus location
    M_septal[12:14, 10:12] = 3.0
    masks['left'] = M_septal

    # Lateral mitral annulus (left-biased L-shape)
    M_lateral = np.zeros((size, size), dtype=np.float32)
    # Vertical bar (lateral wall)
    M_lateral[8:14, 2:6] = 1.0
    # Horizontal bar (valve plane)
    M_lateral[12:14, 2:10] = 1.0
    # Focal point at annulus location
    M_lateral[12:14, 4:6] = 3.0
    masks['right'] = M_lateral

    return masks

# Visualize masks with anatomical annotations
def visualize_masks(masks):
    fig, axs = plt.subplots(1, 3, figsize=(15, 5))
    titles = ['Cardiac Apex (V-Shape)', 'Septal Mitral Annulus (L-Shape)', 'Lateral Mitral Annulus (L-Shape)']
    colors = ['red', 'green', 'blue']
    
    for i, (key, mask) in enumerate(masks.items()):
        ax = axs[i]
        im = ax.imshow(mask, cmap='viridis', interpolation='nearest')
        ax.set_title(titles[i], fontsize=12, fontweight='bold', color=colors[i])
        
        # Add annotations for focal points
        if key == 'top':
            ax.text(8, 14, 'Focal Point\n(Apex Tip)', 
                    ha='center', va='center', color='white', 
                    bbox=dict(facecolor='red', alpha=0.7))
        elif key == 'left':
            ax.text(12, 13, 'Focal Point\n(Annulus)', 
                    ha='center', va='center', color='white',
                    bbox=dict(facecolor='green', alpha=0.7))
        elif key == 'right':
            ax.text(4, 13, 'Focal Point\n(Annulus)', 
                    ha='center', va='center', color='white',
                    bbox=dict(facecolor='blue', alpha=0.7))
        
        ax.grid(True, which='both', color='white', linestyle='--', linewidth=0.5)
        fig.colorbar(im, ax=ax, shrink=0.7)
    
    plt.tight_layout()
    plt.savefig('anatomical_mask_designs.png', dpi=300, bbox_inches='tight')
    plt.show()

# Enhanced detection with anatomical constraints
def detect_landmarks(img, windows, masks):
    # Convert to grayscale and normalize
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY).astype(np.float32) / 255.0
    landmarks = {}
    scores = {}
    
    # Create figure for debug visualization
    fig, axs = plt.subplots(3, 4, figsize=(18, 12))
    fig.suptitle('Cardiac Landmark Detection Analysis', fontsize=16, fontweight='bold')
    
    # Confidence threshold
    MIN_CONFIDENCE = 0.5
    
    for idx, key in enumerate(['top', 'left', 'right']):
        x0, y0, w, h = windows[key]
        
        # Extract ROI
        roi = gray[y0:y0+h, x0:x0+w]
        
        # Perform template matching
        result = cv2.matchTemplate(roi, masks[key], cv2.TM_CCOEFF_NORMED)
        
        # Apply anatomical constraints
        constraint = np.ones_like(result)
        if key == 'top':  # Apex should be in bottom half
            constraint[:h//2, :] = 0
        elif key == 'left':  # Septal annulus in bottom right quadrant
            constraint[:h//2, :] = 0
            constraint[:, :w//2] = 0
        elif key == 'right':  # Lateral annulus in bottom left quadrant
            constraint[:h//2, :] = 0
            constraint[:, w//2:] = 0
        
        # Apply constraint
        constrained_result = result * constraint
        
        # Find best match location
        _, max_val, _, max_loc = cv2.minMaxLoc(constrained_result)
        dx, dy = max_loc
        
        # Calculate landmark position (center of mask)
        x_center = x0 + dx + MASK_SIZE[1]//2
        y_center = y0 + dy + MASK_SIZE[0]//2
        
        # Apply confidence threshold
        if max_val < MIN_CONFIDENCE:
            print(f"Warning: Low confidence ({max_val:.2f}) for {key}, using fallback position")
            # Fallback to window center
            x_center = x0 + w//2
            y_center = y0 + h//2
            max_val = 0.0  # Reset confidence
        
        landmarks[key] = (x_center, y_center)
        scores[key] = max_val
        
        # --- Debug Visualizations ---
        
        # ROI Visualization
        ax = axs[idx, 0]
        ax.imshow(roi, cmap='gray')
        ax.set_title(f'{key.upper()} ROI', fontsize=12)
        
        # Mask Visualization
        ax = axs[idx, 1]
        im = ax.imshow(masks[key], cmap='viridis', interpolation='nearest')
        ax.set_title(f'{key.upper()} Mask', fontsize=12)
        
        # Constraint Visualization
        ax = axs[idx, 2]
        ax.imshow(constraint, cmap='gray', vmin=0, vmax=1)
        ax.set_title(f'{key.upper()} Search Constraint', fontsize=12)
        
        # Matching Result Visualization
        ax = axs[idx, 3]
        im = ax.imshow(constrained_result, cmap='plasma', vmin=-1, vmax=1)
        ax.plot(dx, dy, 'ro', markersize=8, fillstyle='none', linewidth=2)
        ax.set_title(f'Constrained Match (Score: {max_val:.2f})', fontsize=12)
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.savefig('landmark_detection_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return landmarks, scores

# Enhanced visualization of results
def visualize_results(img, windows, landmarks, scores):
    result_img = img.copy()
    colors = {'top': (255, 50, 50), 'left': (50, 255, 50), 'right': (50, 100, 255)}
    labels = {
        'top': 'Cardiac Apex',
        'left': 'Septal Mitral Annulus',
        'right': 'Lateral Mitral Annulus'
    }
    
    # Draw search windows and landmarks
    for key in windows:
        x0, y0, w, h = windows[key]
        color = colors[key]
        
        # Draw search window
        cv2.rectangle(result_img, (x0, y0), (x0+w, y0+h), color, 2)
        
        # Draw landmark
        lx, ly = landmarks[key]
        cv2.circle(result_img, (int(lx), int(ly)), 8, color, -1)
        
        # Draw crosshair at landmark
        cv2.drawMarker(result_img, (int(lx), int(ly)), (255, 255, 255), 
                       markerType=cv2.MARKER_CROSS, markerSize=20, thickness=2)
        
        # Display information
        info = f"{labels[key]}: ({int(lx)}, {int(ly)})"
        cv2.putText(result_img, info, (int(lx)+15, int(ly)-15), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
        cv2.putText(result_img, info, (int(lx)+15, int(ly)-15), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)
        
        # Display confidence
        conf = f"Confidence: {scores[key]:.2f}"
        cv2.putText(result_img, conf, (int(lx)+15, int(ly)+5), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)
        cv2.putText(result_img, conf, (int(lx)+15, int(ly)+5), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)
    
    # Create figure for display
    plt.figure(figsize=(10, 8))
    plt.imshow(result_img)
    plt.title('Cardiac Landmark Detection Results', fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.tight_layout()
    plt.savefig('cardiac_landmark_detection.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return result_img

# Print results in a formatted table
def print_results(landmarks, scores):
    print("\n" + "="*60)
    print("CARDIAC LANDMARK DETECTION RESULTS")
    print("="*60)
    print(f"{'Landmark':<25} | {'Position (x,y)':<12} | {'Confidence':<10} | {'Status':<10}")
    print("-"*60)
    
    for key in landmarks:
        x, y = landmarks[key]
        landmark_name = {
            'top': 'Cardiac Apex',
            'left': 'Septal Mitral Annulus',
            'right': 'Lateral Mitral Annulus'
        }[key]
        
        status = "SUCCESS" if scores[key] >= 0.5 else "LOW CONFIDENCE"
        print(f"{landmark_name:<25} | ({int(x):3d}, {int(y):3d})   | {scores[key]:<10.4f} | {status:<10}")
    
    print("="*60)

# Execute pipeline
print("Creating anatomical masks...")
masks = create_anatomical_masks()
visualize_masks(masks)

print("Detecting landmarks with anatomical constraints...")
landmarks, scores = detect_landmarks(img, windows, masks)

print("Visualizing results...")
result_img = visualize_results(img, windows, landmarks, scores)

print("Final results:")
print_results(landmarks, scores)

# Save final result image
cv2.imwrite('final_detection_result.png', cv2.cvtColor(result_img, cv2.COLOR_RGB2BGR))
print("Results saved to 'final_detection_result.png'")
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Load and prepare image
data = np.load("C:\\Users\\Mesterpc.ir\\Downloads\\masks.npy")
img = cv2.cvtColor(data[1, :, :, :], cv2.COLOR_BGR2RGB)

# Parameters
MASK_SIZE = (16, 16)  # 16x16 pixels as specified
windows = {
    'top': (105, 21, 27, 39),    # (x, y, width, height)
    'left': (89, 126, 46, 57),   # Septal mitral annulus
    'right': (130, 119, 46, 57)  # Lateral mitral annulus
}

# Create precise anatomical masks
def create_precise_masks():
    masks = {}
    size = 16
    
    # Cardiac apex mask (focused V-shape)
    M_top = np.zeros((size, size), dtype=np.float32)
    # Create sharp V-shape
    for i in range(size):
        # Narrow at top, widest at bottom
        width = int(1 + 8 * (i/(size-1)))
        start = max(0, (size//2) - width//2)
        end = min(size, (size//2) + width//2 + 1)
        M_top[i, start:end] = 1.0
    # Add focal point at bottom center
    M_top[12:16, 6:10] = 3.0
    masks['top'] = M_top

    # Mitral annuli masks with precise focal points
    M_septal = np.zeros((size, size), dtype=np.float32)
    # Septal mask: focal point at expected location
    M_septal[8:12, 8:12] = 3.0  # Bright focal point
    M_septal[6:14, 6:14] = 1.0  # Surrounding area
    masks['left'] = M_septal

    M_lateral = np.zeros((size, size), dtype=np.float32)
    # Lateral mask: focal point at expected location
    M_lateral[8:12, 4:8] = 3.0  # Bright focal point
    M_lateral[6:14, 2:10] = 1.0  # Surrounding area
    masks['right'] = M_lateral

    return masks

# Create weighting maps for each window
def create_weight_maps():
    weights = {}
    
    # Weight map for top window (focus on bottom center)
    w_top = np.zeros((39, 27), dtype=np.float32)
    for y in range(39):
        for x in range(27):
            # Higher weight toward bottom center
            dist_y = abs(y - 30) / 39.0
            dist_x = abs(x - 13.5) / 27.0
            w_top[y, x] = 1.0 - 0.7 * (dist_x + dist_y)
    weights['top'] = np.clip(w_top, 0.3, 1.0)
    
    # Weight map for left window (focus on specific point)
    w_left = np.zeros((57, 46), dtype=np.float32)
    # Focus around (25, 8) in ROI coordinates
    for y in range(57):
        for x in range(46):
            dist = np.sqrt((x-25)**2 + (y-8)**2) / np.sqrt(46**2 + 57**2)
            w_left[y, x] = 1.0 - 0.8 * dist
    weights['left'] = np.clip(w_left, 0.2, 1.0)
    
    # Weight map for right window (focus on specific point)
    w_right = np.zeros((57, 46), dtype=np.float32)
    # Focus around (25, 8) in ROI coordinates
    for y in range(57):
        for x in range(46):
            dist = np.sqrt((x-25)**2 + (y-8)**2) / np.sqrt(46**2 + 57**2)
            w_right[y, x] = 1.0 - 0.8 * dist
    weights['right'] = np.clip(w_right, 0.2, 1.0)
    
    return weights

# Enhanced detection with weighted matching
def detect_landmarks(img, windows, masks, weights):
    # Convert to grayscale and normalize
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY).astype(np.float32) / 255.0
    landmarks = {}
    scores = {}
    
    for key in ['top', 'left', 'right']:
        x0, y0, w, h = windows[key]
        mask = masks[key]
        weight_map = weights[key]
        
        # Extract ROI
        roi = gray[y0:y0+h, x0:x0+w]
        
        # Apply preprocessing - enhance contrast
        roi_norm = (roi - np.min(roi)) / (np.max(roi) - np.min(roi) + 1e-7)
        
        # Perform weighted template matching
        result = cv2.matchTemplate(roi_norm, mask, cv2.TM_CCOEFF_NORMED)
        
        # Apply weighting to the result map
        weighted_result = result * weight_map[:result.shape[0], :result.shape[1]]
        
        # Find best match location
        _, max_val, _, max_loc = cv2.minMaxLoc(weighted_result)
        dx, dy = max_loc
        
        # Calculate landmark position (center of mask)
        x_center = x0 + dx + MASK_SIZE[1]//2
        y_center = y0 + dy + MASK_SIZE[0]//2
        
        landmarks[key] = (x_center, y_center)
        scores[key] = max_val
        
        # Debug visualization
        plt.figure(figsize=(15, 4))
        
        # Show ROI
        plt.subplot(141)
        plt.imshow(roi, cmap='gray')
        plt.title(f'{key.upper()} ROI')
        
        # Show mask
        plt.subplot(142)
        plt.imshow(mask, cmap='viridis')
        plt.title(f'{key.upper()} Mask')
        
        # Show weight map
        plt.subplot(143)
        plt.imshow(weight_map, cmap='hot')
        plt.title('Weight Map')
        
        # Show matching result
        plt.subplot(144)
        plt.imshow(weighted_result, cmap='plasma')
        plt.plot(dx, dy, 'ro', markersize=8, fillstyle='none')
        plt.title(f'Weighted Match (Score: {max_val:.2f})')
        plt.colorbar()
        
        plt.tight_layout()
        plt.savefig(f'{key}_detection_debug.png', dpi=300)
        plt.show()
    
    return landmarks, scores

# Enhanced visualization of results
def visualize_results(img, windows, landmarks, scores):
    result_img = img.copy()
    colors = {'top': (255, 50, 50), 'left': (50, 255, 50), 'right': (50, 100, 255)}
    labels = {
        'top': 'Cardiac Apex',
        'left': 'Septal Mitral Annulus',
        'right': 'Lateral Mitral Annulus'
    }
    
    # Draw search windows and landmarks
    for key in windows:
        x0, y0, w, h = windows[key]
        color = colors[key]
        
        # Draw search window
        cv2.rectangle(result_img, (x0, y0), (x0+w, y0+h), color, 2)
        
        # Draw landmark
        lx, ly = landmarks[key]
        cv2.circle(result_img, (int(lx), int(ly)), 10, color, -1)
        
        # Draw crosshair at landmark
        cv2.drawMarker(result_img, (int(lx), int(ly)), (255, 255, 255), 
                       markerType=cv2.MARKER_CROSS, markerSize=25, thickness=2)
        
        # Display information
        info = f"{labels[key]}: ({int(lx)}, {int(ly)})"
        cv2.putText(result_img, info, (int(lx)+20, int(ly)-20), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
        cv2.putText(result_img, info, (int(lx)+20, int(ly)-20), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 1)
        
        # Display confidence
        conf = f"Confidence: {scores[key]:.2f}"
        cv2.putText(result_img, conf, (int(lx)+20, int(ly)+10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
        cv2.putText(result_img, conf, (int(lx)+20, int(ly)+10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 1)
    
    # Create figure for display
    plt.figure(figsize=(10, 8))
    plt.imshow(result_img)
    plt.title('Precise Cardiac Landmark Detection', fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.tight_layout()
    plt.savefig('precise_landmark_detection.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    return result_img

# Print results in a formatted table
def print_results(landmarks, scores):
    print("\n" + "="*60)
    print("PRECISE CARDIAC LANDMARK DETECTION RESULTS")
    print("="*60)
    print(f"{'Landmark':<25} | {'Position (x,y)':<12} | {'Confidence':<10}")
    print("-"*60)
    
    for key in landmarks:
        x, y = landmarks[key]
        landmark_name = {
            'top': 'Cardiac Apex',
            'left': 'Septal Mitral Annulus',
            'right': 'Lateral Mitral Annulus'
        }[key]
        
        print(f"{landmark_name:<25} | ({int(x):3d}, {int(y):3d})   | {scores[key]:<10.4f}")
    
    print("="*60)

# Execute pipeline
print("Creating precise masks...")
masks = create_precise_masks()

print("Creating weight maps...")
weights = create_weight_maps()

print("Detecting landmarks with weighted matching...")
landmarks, scores = detect_landmarks(img, windows, masks, weights)

print("Visualizing results...")
result_img = visualize_results(img, windows, landmarks, scores)

print("Final results:")
print_results(landmarks, scores)

# Save final result image
cv2.imwrite('precise_detection_result.png', cv2.cvtColor(result_img, cv2.COLOR_RGB2BGR))
print("Results saved to 'precise_detection_result.png'")
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Load and prepare image
data = np.load("C:\\Users\\Mesterpc.ir\\Downloads\\masks.npy")
img = cv2.cvtColor(data[1, :, :, :], cv2.COLOR_BGR2RGB)

# Parameters
MASK_SIZE = (16, 16)
windows = {
    'top': (105, 21, 27, 39),
    'left': (89, 126, 46, 57),   # Septal mitral annulus
    'right': (130, 119, 46, 57)  # Lateral mitral annulus
}

# Create precise anatomical masks with upper focal points
def create_precise_masks():
    masks = {}
    size = 16
    
    # Cardiac apex mask
    M_top = np.zeros((size, size), dtype=np.float32)
    for i in range(size):
        width = int(1 + 8 * (i/(size-1)))
        start = max(0, (size//2) - width//2)
        end = min(size, (size//2) + width//2 + 1)
        M_top[i, start:end] = 1.0
    M_top[12:16, 6:10] = 3.0
    masks['top'] = M_top

    # Mitral annuli masks with UPPER focal points
    # Left mask (septal): focal point in upper portion
    M_septal = np.zeros((size, size), dtype=np.float32)
    M_septal[4:12, 4:12] = 1.0  # Surrounding area
    M_septal[4:8, 6:10] = 3.0    # Bright focal point
    masks['left'] = M_septal

    # Right mask (lateral): modified to be less specific
    M_lateral = np.zeros((size, size), dtype=np.float32)
    # Broader area with lower intensity
    M_lateral[2:14, 2:14] = 0.5  # Weaker surrounding
    M_lateral[4:8, 4:8] = 1.0    # Moderate focal point
    masks['right'] = M_lateral

    return masks

# Create weighting maps with 0.1-0.2 normalized y-coordinates
def create_weight_maps():
    weights = {}
    
    # Weight map for top window
    w_top = np.zeros((39, 27), dtype=np.float32)
    for y in range(39):
        for x in range(27):
            dist_y = abs(y - 30) / 39.0
            dist_x = abs(x - 13.5) / 27.0
            w_top[y, x] = 1.0 - 0.7 * (dist_x + dist_y)
    weights['top'] = np.clip(w_top, 0.3, 1.0)
    
    # Left window: focal point at y=6 (6/570.105)
    w_left = np.zeros((57, 46), dtype=np.float32)
    focus_y_left = int(57 * 0.105)  # 6
    for y in range(57):
        for x in range(46):
            dist = np.sqrt((x-25)**2 + (y-focus_y_left)**2) / np.sqrt(46**2 + 57**2)
            w_left[y, x] = 1.0 - 0.8 * dist
    weights['left'] = np.clip(w_left, 0.2, 1.0)
    
    # Right window: focal point at y=11 (11/570.193)
    w_right = np.zeros((57, 46), dtype=np.float32)
    focus_y_right = int(57 * 0.193)  # 11
    # Broader focus area with lower weights
    for y in range(57):
        for x in range(46):
            dist = np.sqrt((x-25)**2 + (y-focus_y_right)**2) / np.sqrt(46**2 + 57**2)
            w_right[y, x] = 1.0 - 1.2 * dist  # Steeper falloff for lower scores
    weights['right'] = np.clip(w_right, 0.1, 0.8)  # Cap max weight

    return weights

# Enhanced detection with weighted matching
def detect_landmarks(img, windows, masks, weights):
    # Convert to grayscale and normalize
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY).astype(np.float32) / 255.0
    landmarks = {}
    scores = {}
    
    for key in ['top', 'left', 'right']:
        x0, y0, w, h = windows[key]
        mask = masks[key]
        weight_map = weights[key]
        
        # Extract ROI
        roi = gray[y0:y0+h, x0:x0+w]
        
        # Apply preprocessing - enhance contrast
        roi_norm = (roi - np.min(roi)) / (np.max(roi) - np.min(roi) + 1e-7)
        
        # Perform weighted template matching
        result = cv2.matchTemplate(roi_norm, mask, cv2.TM_CCOEFF_NORMED)
        
        # Apply weighting to the result map
        weighted_result = result * weight_map[:result.shape[0], :result.shape[1]]
        
        # Find best match location
        _, max_val, _, max_loc = cv2.minMaxLoc(weighted_result)
        dx, dy = max_loc
        
        # Calculate landmark position (center of mask)
        x_center = x0 + dx + MASK_SIZE[1]//2
        y_center = y0 + dy + MASK_SIZE[0]//2
        
        landmarks[key] = (x_center, y_center)
        scores[key] = max_val
        
        # Debug visualization
        plt.figure(figsize=(15, 4))
        
        # Show ROI
        plt.subplot(141)
        plt.imshow(roi, cmap='gray')
        plt.title(f'{key.upper()} ROI')
        
        # Show mask
        plt.subplot(142)
        plt.imshow(mask, cmap='viridis')
        plt.title(f'{key.upper()} Mask')  # FIXED TYPO HERE: .upper() not .UPER()
        
        # Show weight map
        plt.subplot(143)
        plt.imshow(weight_map, cmap='hot')
        plt.title('Weight Map')
        
        # Show matching result
        plt.subplot(144)
        plt.imshow(weighted_result, cmap='plasma')
        plt.plot(dx, dy, 'ro', markersize=8, fillstyle='none')
        plt.title(f'Weighted Match (Score: {max_val:.2f})')
        plt.colorbar()
        
        plt.tight_layout()
        plt.savefig(f'{key}_detection_debug.png', dpi=300)
        plt.show()
    
    return landmarks, scores

# Enhanced visualization of results
def visualize_results(img, windows, landmarks, scores):
    result_img = img.copy()
    colors = {
        'top': (0, 0, 255),      # Red - Cardiac apex
        'left': (0, 255, 0),     # Green - Septal mitral annulus
        'right': (255, 0, 0)     # Blue - Lateral mitral annulus
    }
    
    # Draw landmarks with the specified style
    for key in landmarks:
        lx, ly = landmarks[key]
        color = colors[key]
        
        # Draw larger circle with outline
        cv2.circle(result_img, (int(lx), int(ly)), 10, (255, 255, 255), 3)  # White outline
        cv2.circle(result_img, (int(lx), int(ly)), 10, color, -1)           # Filled color center
        
        # Draw crosshair marker
        cv2.drawMarker(result_img, (int(lx), int(ly)), (255, 255, 255), 
                       markerType=cv2.MARKER_CROSS, markerSize=30, thickness=2)
        
        # Prepare text labels based on landmark type
        if key == 'top':
            label = f"Cardiac Apex: ({int(lx)}, {int(ly)})"
            conf_label = f"Confidence: {scores[key]:.2f}"
        elif key == 'left':
            label = f"Septal Mitral Annulus: ({int(lx)}, {int(ly)})"
            conf_label = f"Confidence: {scores[key]:.2f}"
        else:  # right
            label = f"Lateral Mitral Annulus: ({int(lx)}, {int(ly)})"
            conf_label = f"Confidence: {scores[key]:.2f}"
        
        # Position text based on landmark location
        text_x = int(lx) + 20
        if key == 'top':
            text_y = int(ly) - 30
        elif key == 'left':
            text_y = int(ly) - 40
        else:  # right
            text_y = int(ly) + 40
        
        # Draw text with background for readability
        cv2.putText(result_img, label, (text_x, text_y), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 0), 4)  # Black background
        cv2.putText(result_img, label, (text_x, text_y), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)  # White text
        
        # Draw confidence slightly below
        cv2.putText(result_img, conf_label, (text_x, text_y + 30), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 0), 4)  # Black background
        cv2.putText(result_img, conf_label, (text_x, text_y + 30), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)  # White text
    
    # Display and save
    plt.figure(figsize=(12, 10))
    plt.imshow(result_img)
    plt.axis('off')
    plt.tight_layout()
    plt.savefig('precise_landmark_detection.png', dpi=300, bbox_inches='tight', pad_inches=0.1)
    plt.show()
    
    return result_img

# Print results in a formatted table
def print_results(landmarks, scores):
    print("\n" + "="*60)
    print("PRECISE CARDIAC LANDMARK DETECTION RESULTS")
    print("="*60)
    print(f"{'Landmark':<25} | {'Position (x,y)':<12} | {'Confidence':<10}")
    print("-"*60)
    
    for key in landmarks:
        x, y = landmarks[key]
        landmark_name = {
            'top': 'Cardiac Apex',
            'left': 'Septal Mitral Annulus',
            'right': 'Lateral Mitral Annulus'
        }[key]
        
        print(f"{landmark_name:<25} | ({int(x):3d}, {int(y):3d})   | {scores[key]:<10.4f}")
    
    print("="*60)

# Execute pipeline
print("Creating precise masks with upper focal points...")
masks = create_precise_masks()

print("Creating weight maps with 0.1-0.2 normalized y-coordinates...")
weights = create_weight_maps()

print("Detecting landmarks with weighted matching...")
landmarks, scores = detect_landmarks(img, windows, masks, weights)

print("Visualizing results...")
result_img = visualize_results(img, windows, landmarks, scores)

print("Final results:")
print_results(landmarks, scores)

# Save final result image
cv2.imwrite('precise_detection_result.png', cv2.cvtColor(result_img, cv2.COLOR_RGB2BGR))
print("Results saved to 'precise_detection_result.png'")
```
