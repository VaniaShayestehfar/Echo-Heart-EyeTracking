```python
import os
import numpy as np
import nibabel as nib
import cv2
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
import math

def rotate_image(image, angle=90):
    """Rotate image by specified angle (degrees)"""
    if angle == 0:
        return image
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    return cv2.warpAffine(image, M, (w, h))

def detect_landmarks(seg_binary, delta_y_fraction=0.1, y_tolerance=10):
    """Detect apex, septal, and lateral landmarks with curvature-based refinement."""
    seg_binary = (seg_binary > 0).astype(np.uint8) * 255
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        raise ValueError("No LV contour found")
    
    contour = max(contours, key=cv2.contourArea).squeeze()
    x = contour[:, 0].astype(float)
    y = contour[:, 1].astype(float)
    
    # Smooth contour
    x_smooth = gaussian_filter1d(x, sigma=5, mode='wrap')
    y_smooth = gaussian_filter1d(y, sigma=5, mode='wrap')
    
    # Compute curvature
    dx = np.gradient(x_smooth)
    dy = np.gradient(y_smooth)
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(dx * ddy - dy * ddx) / (dx**2 + dy**2 + 1e-10)**1.5
    curvature = (curvature - np.min(curvature)) / (np.max(curvature) - np.min(curvature) + 1e-10)
    
    # Find apex: min y
    apex_idx = np.argmin(y)
    apex_point = (int(x[apex_idx]), int(y[apex_idx]))
    
    # Find all peaks in curvature
    peaks, _ = find_peaks(curvature, prominence=0.0005)
    
    # Define basal region
    max_y = max(y)
    min_y = min(y)
    basal_threshold = max_y - delta_y_fraction * (max_y - min_y)
    basal_peaks = [i for i in peaks if y[i] >= basal_threshold]
    
    if len(basal_peaks) >= 2:
        sorted_peaks = sorted(basal_peaks, key=lambda i: curvature[i], reverse=True)
        septal_idx = sorted_peaks[0]
        lateral_idx = sorted_peaks[1]
        
        if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
            septal_y = y[septal_idx]
            lateral_candidates = [i for i in basal_peaks if abs(y[i] - septal_y) <= y_tolerance]
            if len(lateral_candidates) >= 2:
                lateral_candidates = [i for i in lateral_candidates if i != septal_idx]
                if lateral_candidates:
                    lateral_idx = max(lateral_candidates, key=lambda i: curvature[i])
                else:
                    raise ValueError("No suitable lateral peak found")
            else:
                raise ValueError("Not enough peaks with similar y-coordinates")
    else:
        mean_y = np.mean(y)
        lower_indices = [i for i in range(len(y)) if y[i] > mean_y]
        if len(lower_indices) >= 2:
            septal_idx = min(lower_indices, key=lambda i: x[i])
            lateral_idx = max(lower_indices, key=lambda i: x[i])
            if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
                raise ValueError("Fallback points not aligned")
        else:
            raise ValueError("Not enough points for fallback")
    
    septal_point = (int(x[septal_idx]), int(y[septal_idx]))
    lateral_point = (int(x[lateral_idx]), int(y[lateral_idx]))
    mitral_mid = ((septal_point[0] + lateral_point[0]) // 2, (septal_point[1] + lateral_point[1]) // 2)
    
    return apex_point, septal_point, lateral_point, mitral_mid

def compute_disk_diameters(seg_binary, apex, mitral_mid, pixel_spacing=(0.1, 0.1), nr_disks=20):
    """Compute disk diameters and long axis length."""
    contours, _ = cv2.findContours(seg_binary.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        return 0.0, [], 0.0
    
    max_contour = max(contours, key=cv2.contourArea)
    filled_mask = np.zeros_like(seg_binary)
    cv2.drawContours(filled_mask, [max_contour], -1, 1, thickness=cv2.FILLED)
    
    kernel = np.ones((3,3), np.uint8)
    dilation = cv2.dilate(filled_mask, kernel)
    erosion = cv2.erode(filled_mask, kernel)
    boundary = (dilation - erosion > 0).astype(np.uint8)
    
    # Long axis vector
    dx = apex[0] - mitral_mid[0]
    dy = apex[1] - mitral_mid[1]
    L_px = np.sqrt(dx**2 + dy**2)
    L_cm = L_px * pixel_spacing[0]
    
    if L_cm == 0:
        return L_cm, [], 0.0
    
    perp_vector = np.array([-dy, dx]) / L_px if L_px > 0 else np.array([1, 0])
    
    diameters_cm = []
    disk_info = []
    
    for i in range(nr_disks):
        t = (i + 0.5) / nr_disks
        center_x = mitral_mid[0] + t * dx
        center_y = mitral_mid[1] + t * dy
        
        left_point = None
        right_point = None
        
        for direction in [-1, 1]:
            for dist in range(1, 150):
                px = int(round(center_x + direction * dist * perp_vector[0]))
                py = int(round(center_y + direction * dist * perp_vector[1]))
                
                if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                    break
                
                if boundary[py, px]:
                    if direction == -1:
                        left_point = (px, py)
                    else:
                        right_point = (px, py)
                    break
        
        if left_point and right_point:
            dx_px = right_point[0] - left_point[0]
            dy_px = right_point[1] - left_point[1]
            diameter_cm = np.sqrt((dx_px * pixel_spacing[0])**2 + (dy_px * pixel_spacing[1])**2)
            diameters_cm.append(diameter_cm)
            disk_info.append({
                'center': (center_x, center_y),
                'left': left_point,
                'right': right_point,
                'diameter_cm': diameter_cm
            })
        else:
            diameters_cm.append(0.0)
    
    # Calculate dead space area
    dead_area_cm2 = 0.0
    if disk_info:
        polygon_points = []
        for disk in disk_info:
            if disk['left']:
                polygon_points.append(disk['left'])
        for disk in reversed(disk_info):
            if disk['right']:
                polygon_points.append(disk['right'])
        
        if len(polygon_points) > 2:
            disk_area_mask = np.zeros_like(filled_mask, dtype=np.uint8)
            pts = np.array([polygon_points], dtype=np.int32)
            cv2.fillPoly(disk_area_mask, [pts], 1)
            dead_space_mask = filled_mask - disk_area_mask
            dead_space_mask = np.clip(dead_space_mask, 0, 1)
            dead_area_px = np.sum(dead_space_mask)
            dead_area_cm2 = dead_area_px * pixel_spacing[0] * pixel_spacing[1]
    
    return L_cm, diameters_cm, dead_area_cm2

def compute_biplane_volume(L4_cm, L2_cm, diameters4_cm, diameters2_cm, dead_area4_cm2, dead_area2_cm2, nr_disks=20):
    """Compute biplane Simpson's volume with dead space correction."""
    L_avg_cm = (L4_cm + L2_cm) / 2
    h_cm = L_avg_cm / nr_disks
    
    total_volume = 0.0
    for i in range(nr_disks):
        a_i = diameters4_cm[i]
        b_i = diameters2_cm[i]
        disk_volume = (math.pi / 4) * a_i * b_i * h_cm
        total_volume += disk_volume
    
    avg_dead_area_cm2 = (dead_area4_cm2 + dead_area2_cm2) / 2
    dead_volume = avg_dead_area_cm2 * h_cm
    total_volume_with_dead = total_volume + dead_volume
    
    return total_volume_with_dead

def process_patient(patient_dir):
    """Process all images for a single patient."""
    volumes = {'2CH': {'ED': None, 'ES': None}, '4CH': {'ED': None, 'ES': None}}
    
    for file in os.listdir(patient_dir):
        if file.endswith('_gt.nii.gz'):
            # Determine view and phase
            view = '4CH' if '4ch' in file.lower() else '2CH' if '2ch' in file.lower() else None
            phase = 'ED' if 'ed' in file.lower() else 'ES' if 'es' in file.lower() else None
            
            if not view or not phase:
                continue
            
            gt_path = os.path.join(patient_dir, file)
            gt_img = nib.load(gt_path)
            gt_data = gt_img.get_fdata()
            pixel_spacing = gt_img.header.get_zooms()[:2]
            
            # Create LV mask
            mask = np.zeros_like(gt_data, dtype=np.uint8)
            mask[gt_data == 1] = 255
            
            # Rotate 90 degrees
            mask = rotate_image(mask, 90)
            
            try:
                apex, septal, lateral, mitral_mid = detect_landmarks(mask)
                L_cm, diameters_cm, dead_area_cm2 = compute_disk_diameters(
                    mask, apex, mitral_mid, pixel_spacing=pixel_spacing
                )
                volumes[view][phase] = {
                    'L_cm': L_cm,
                    'diameters': diameters_cm,
                    'dead_area': dead_area_cm2
                }
            except Exception as e:
                print(f"Error processing {file}: {str(e)}")
                continue
    
    return volumes

def calculate_ef(patient_volumes):
    """Calculate ejection fraction from processed volumes."""
    if (patient_volumes['4CH']['ED'] and patient_volumes['2CH']['ED'] and 
        patient_volumes['4CH']['ES'] and patient_volumes['2CH']['ES']):
        
        # Calculate ED volume
        ed_vol = compute_biplane_volume(
            patient_volumes['4CH']['ED']['L_cm'],
            patient_volumes['2CH']['ED']['L_cm'],
            patient_volumes['4CH']['ED']['diameters'],
            patient_volumes['2CH']['ED']['diameters'],
            patient_volumes['4CH']['ED']['dead_area'],
            patient_volumes['2CH']['ED']['dead_area']
        )
        
        # Calculate ES volume
        es_vol = compute_biplane_volume(
            patient_volumes['4CH']['ES']['L_cm'],
            patient_volumes['2CH']['ES']['L_cm'],
            patient_volumes['4CH']['ES']['diameters'],
            patient_volumes['2CH']['ES']['diameters'],
            patient_volumes['4CH']['ES']['dead_area'],
            patient_volumes['2CH']['ES']['dead_area']
        )
        
        # Calculate EF
        sv = ed_vol - es_vol
        ef = (sv / ed_vol) * 100 if ed_vol > 0 else 0
        
        return {
            'EDV': ed_vol,
            'ESV': es_vol,
            'SV': sv,
            'EF': ef
        }
    return None

# Main processing
base_dir = r"E:\echo heart\Resources"
results = []

for patient_dir in [os.path.join(base_dir, d) for d in os.listdir(base_dir) 
                   if os.path.isdir(os.path.join(base_dir, d))]:
    patient_id = os.path.basename(patient_dir)
    print(f"\nProcessing {patient_id}")
    
    # Process all images for this patient
    patient_volumes = process_patient(patient_dir)
    
    # Calculate EF if we have all required data
    ef_result = calculate_ef(patient_volumes)
    
    if ef_result:
        results.append({
            'patient_id': patient_id,
            **ef_result
        })
    else:
        print(f"Incomplete data for {patient_id} - skipping EF calculation")

# Print results
print("\n" + "="*50)
print("LV Volume and Ejection Fraction Results")
print("="*50 + "\n")

for res in results:
    print(f"Patient: {res['patient_id']}")
    print(f"  EDV: {res['EDV']:.2f} ml")
    print(f"  ESV: {res['ESV']:.2f} ml")
    print(f"  SV: {res['SV']:.2f} ml")
    print(f"  EF: {res['EF']:.2f} %")
    print("-"*40)

# Visualize EF distribution
if results:
    ef_values = [res['EF'] for res in results]
    plt.figure(figsize=(10, 6))
    plt.hist(ef_values, bins=15, color='skyblue', edgecolor='black')
    plt.axvline(np.mean(ef_values), color='red', linestyle='dashed', linewidth=1, 
               label=f'Mean EF: {np.mean(ef_values):.1f}%')
    plt.title('Ejection Fraction Distribution')
    plt.xlabel('Ejection Fraction (%)')
    plt.ylabel('Number of Patients')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
else:
    print("No complete patient datasets for EF calculation")
```

```python
import os
import numpy as np
import nibabel as nib
import cv2
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
import math

def rotate_image(image, angle=90):
    """Rotate image by specified angle (degrees) with visualization"""
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, -angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h))
    
    # Visualization
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(image, cmap='gray')
    plt.title("Original Mask")
    plt.axis('off')
    
    plt.subplot(122)
    plt.imshow(rotated, cmap='gray')
    plt.title(f"Rotated ({angle})")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    return rotated

def visualize_mask_creation(gt_data):
    """Visualize the mask creation process"""
    mask = np.zeros_like(gt_data, dtype=np.uint8)
    mask[gt_data == 1] = 255
    
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(gt_data, cmap='gray')
    plt.title("Original Ground Truth")
    plt.axis('off')
    
    plt.subplot(122)
    plt.imshow(mask, cmap='gray')
    plt.title("LV Mask (White=LV, Black=Background)")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    return mask

def detect_landmarks_with_visualization(seg_binary):
    """Detect landmarks with visualization of each step"""
    seg_binary = (seg_binary > 0).astype(np.uint8) * 255
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    
    if not contours:
        raise ValueError("No LV contour found")
    
    # Create visualization canvas
    vis = np.stack([seg_binary]*3, axis=-1)
    
    # Get largest contour
    contour = max(contours, key=cv2.contourArea).squeeze()
    x = contour[:, 0].astype(float)
    y = contour[:, 1].astype(float)
    
    # Draw contour
    for i in range(len(contour)):
        cv2.circle(vis, tuple(contour[i][0]), 1, (0, 255, 0), -1)
    
    # Smooth contour
    x_smooth = gaussian_filter1d(x, sigma=5, mode='wrap')
    y_smooth = gaussian_filter1d(y, sigma=5, mode='wrap')
    smooth_contour = np.column_stack([x_smooth, y_smooth]).astype(int)
    
    # Compute curvature
    dx = np.gradient(x_smooth)
    dy = np.gradient(y_smooth)
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(dx * ddy - dy * ddx) / (dx**2 + dy**2 + 1e-10)**1.5
    curvature = (curvature - np.min(curvature)) / (np.max(curvature) - np.min(curvature) + 1e-10)
    
    # Find apex (min y)
    apex_idx = np.argmin(y)
    apex_point = (int(x[apex_idx]), int(y[apex_idx]))
    cv2.circle(vis, apex_point, 5, (255, 0, 0), -1)  # Red
    
    # Find peaks in curvature
    peaks, _ = find_peaks(curvature, prominence=0.0005)
    
    # Visualize curvature
    plt.figure(figsize=(15, 5))
    plt.subplot(131)
    plt.plot(curvature)
    plt.title("Curvature Along Contour")
    plt.xlabel("Contour point index")
    plt.ylabel("Normalized curvature")
    
    plt.subplot(132)
    plt.plot(x, y, 'g-', label='Original')
    plt.plot(x_smooth, y_smooth, 'b-', label='Smoothed')
    plt.plot(x[peaks], y[peaks], 'ro', label='Peaks')
    plt.plot(apex_point[0], apex_point[1], 'mo', markersize=10, label='Apex')
    plt.title("Contour with Peaks")
    plt.legend()
    plt.gca().invert_yaxis()
    
    plt.subplot(133)
    plt.imshow(vis)
    plt.title("Detected Contour Points (Green)")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    return apex_point, peaks, x, y, curvature

def compute_disk_diameters_with_visualization(seg_binary, apex, peaks, x, y, pixel_spacing, view, phase):
    """Compute disk diameters with full visualization"""
    # Create filled mask
    seg_binary = (seg_binary > 0).astype(np.uint8)
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    max_contour = max(contours, key=cv2.contourArea)
    filled_mask = np.zeros_like(seg_binary)
    cv2.drawContours(filled_mask, [max_contour], -1, 1, thickness=cv2.FILLED)
    
    # Create boundary
    kernel = np.ones((3,3), np.uint8)
    dilation = cv2.dilate(filled_mask, kernel)
    erosion = cv2.erode(filled_mask, kernel)
    boundary = (dilation - erosion > 0).astype(np.uint8)
    
    # Find septal and lateral points
    max_y = max(y)
    min_y = min(y)
    basal_threshold = max_y - 0.1 * (max_y - min_y)
    basal_peaks = [i for i in peaks if y[i] >= basal_threshold]
    
    if len(basal_peaks) >= 2:
        septal_idx = basal_peaks[0]
        lateral_idx = basal_peaks[1]
    else:
        mean_y = np.mean(y)
        lower_indices = [i for i in range(len(y)) if y[i] > mean_y]
        septal_idx = min(lower_indices, key=lambda i: x[i])
        lateral_idx = max(lower_indices, key=lambda i: x[i])
    
    septal_point = (int(x[septal_idx]), int(y[septal_idx]))
    lateral_point = (int(x[lateral_idx]), int(y[lateral_idx]))
    mitral_mid = ((septal_point[0] + lateral_point[0]) // 2, (septal_point[1] + lateral_point[1]) // 2)
    
    # Long axis
    dx = apex[0] - mitral_mid[0]
    dy = apex[1] - mitral_mid[1]
    L_px = np.sqrt(dx**2 + dy**2)
    L_cm = L_px * pixel_spacing[0]
    
    # Perpendicular vector
    perp_vector = np.array([-dy, dx]) / L_px if L_px > 0 else np.array([1, 0])
    
    # Visualization setup
    vis = np.stack([seg_binary*255]*3, axis=-1)
    cv2.line(vis, apex, mitral_mid, (0, 255, 255), 2)  # Yellow long axis
    cv2.circle(vis, apex, 5, (255, 0, 0), -1)  # Red apex
    cv2.circle(vis, septal_point, 5, (0, 0, 255), -1)  # Blue septal
    cv2.circle(vis, lateral_point, 5, (0, 255, 0), -1)  # Green lateral
    cv2.circle(vis, mitral_mid, 5, (255, 255, 0), -1)  # Cyan mitral mid
    
    nr_disks = 20
    diameters_cm = []
    disk_lines = []
    
    for i in range(nr_disks):
        t = (i + 0.5) / nr_disks
        center_x = mitral_mid[0] + t * dx
        center_y = mitral_mid[1] + t * dy
        
        left_point = None
        right_point = None
        
        # Search left
        for dist in range(1, 150):
            px = int(round(center_x - dist * perp_vector[0]))
            py = int(round(center_y - dist * perp_vector[1]))
            if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                break
            if boundary[py, px]:
                left_point = (px, py)
                break
        
        # Search right
        for dist in range(1, 150):
            px = int(round(center_x + dist * perp_vector[0]))
            py = int(round(center_y + dist * perp_vector[1]))
            if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                break
            if boundary[py, px]:
                right_point = (px, py)
                break
        
        if left_point and right_point:
            dx_px = right_point[0] - left_point[0]
            dy_px = right_point[1] - left_point[1]
            diameter_cm = np.sqrt((dx_px * pixel_spacing[0])**2 + (dy_px * pixel_spacing[1])**2)
            diameters_cm.append(diameter_cm)
            
            # Add to visualization
            cv2.line(vis, left_point, right_point, (255, 0, 255), 1)  # Purple disk line
            disk_lines.append((left_point, right_point))
    
    # Final visualization
    plt.figure(figsize=(15, 5))
    plt.subplot(131)
    plt.imshow(seg_binary, cmap='gray')
    plt.title(f"{view} {phase} - Binary Mask")
    plt.axis('off')
    
    plt.subplot(132)
    plt.imshow(boundary, cmap='gray')
    plt.title("Boundary Detection")
    plt.axis('off')
    
    plt.subplot(133)
    plt.imshow(vis)
    plt.title(f"Landmarks and Disks\nApex(Red), Septal(Blue), Lateral(Green)\nLong Axis: {L_cm:.2f} cm")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    # Calculate dead space area
    dead_area_cm2 = 0.0
    if disk_lines:
        polygon_points = []
        for left, right in disk_lines:
            polygon_points.append(left)
        for left, right in reversed(disk_lines):
            polygon_points.append(right)
        
        if len(polygon_points) > 2:
            disk_area_mask = np.zeros_like(filled_mask, dtype=np.uint8)
            pts = np.array([polygon_points], dtype=np.int32)
            cv2.fillPoly(disk_area_mask, [pts], 1)
            dead_space_mask = filled_mask - disk_area_mask
            dead_space_mask = np.clip(dead_space_mask, 0, 1)
            dead_area_px = np.sum(dead_space_mask)
            dead_area_cm2 = dead_area_px * pixel_spacing[0] * pixel_spacing[1]
            
            # Visualize dead space
            plt.figure(figsize=(10, 5))
            plt.subplot(121)
            plt.imshow(disk_area_mask, cmap='gray')
            plt.title("Disk Area Coverage")
            plt.axis('off')
            
            plt.subplot(122)
            plt.imshow(dead_space_mask, cmap='gray')
            plt.title(f"Dead Space Area: {dead_area_cm2:.2f} cm")
            plt.axis('off')
            plt.tight_layout()
            plt.show()
    
    return L_cm, diameters_cm, dead_area_cm2

def calculate_simpson_volume(L_cm, diameters_cm, dead_area_cm2, nr_disks=20):
    """Calculate volume using Simpson's method with visualization"""
    h_cm = L_cm / nr_disks
    disk_volumes = [(math.pi/4) * d**2 * h_cm for d in diameters_cm]
    total_disk_volume = sum(disk_volumes)
    dead_volume = dead_area_cm2 * h_cm
    total_volume = total_disk_volume + dead_volume
    
    # Visualization
    plt.figure(figsize=(12, 6))
    plt.bar(range(len(disk_volumes)), disk_volumes, color='skyblue', label='Disk Volumes')
    plt.axhline(y=total_disk_volume, color='red', linestyle='--', label=f'Total Disk Volume: {total_disk_volume:.2f} ml')
    plt.axhline(y=dead_volume, color='green', linestyle='--', label=f'Dead Space Volume: {dead_volume:.2f} ml')
    plt.axhline(y=total_volume, color='purple', linestyle='-', label=f'Total Volume: {total_volume:.2f} ml')
    plt.xlabel("Disk Number")
    plt.ylabel("Volume (ml)")
    plt.title("Simpson's Method Volume Calculation")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return total_volume

def process_patient_with_visualization(patient_dir):
    """Process one patient with full visualization of each step"""
    patient_id = os.path.basename(patient_dir)
    print(f"\n{'='*50}")
    print(f"Processing Patient: {patient_id}")
    print(f"{'='*50}\n")
    
    volumes = {'2CH': {'ED': None, 'ES': None}, '4CH': {'ED': None, 'ES': None}}
    
    for file in os.listdir(patient_dir):
        if file.endswith('_gt.nii.gz'):
            # Determine view and phase
            view = '4CH' if '4ch' in file.lower() else '2CH' if '2ch' in file.lower() else None
            phase = 'ED' if 'ed' in file.lower() else 'ES' if 'es' in file.lower() else None
            
            if not view or not phase:
                continue
            
            print(f"\nProcessing {view} {phase} view...")
            
            # Load data
            gt_path = os.path.join(patient_dir, file)
            gt_img = nib.load(gt_path)
            gt_data = gt_img.get_fdata()
            pixel_spacing = gt_img.header.get_zooms()[:2]
            
            # Step 1: Mask creation visualization
            print("\n1. Creating LV Mask...")
            mask = visualize_mask_creation(gt_data)
            
            # Step 2: Rotation visualization
            print("\n2. Rotating Image...")
            rotated_mask = rotate_image(mask, 90)
            
            # Step 3: Landmark detection visualization
            print("\n3. Detecting Landmarks...")
            try:
                apex_point, peaks, x, y, curvature = detect_landmarks_with_visualization(rotated_mask)
                
                # Step 4: Disk diameter calculation visualization
                print("\n4. Calculating Disk Diameters...")
                L_cm, diameters_cm, dead_area_cm2 = compute_disk_diameters_with_visualization(
                    rotated_mask, apex_point, peaks, x, y, pixel_spacing, view, phase
                )
                
                # Step 5: Volume calculation visualization
                print("\n5. Calculating Volume...")
                volume = calculate_simpson_volume(L_cm, diameters_cm, dead_area_cm2)
                
                volumes[view][phase] = {
                    'L_cm': L_cm,
                    'diameters': diameters_cm,
                    'dead_area': dead_area_cm2,
                    'volume': volume
                }
                
                print(f"\n{view} {phase} Volume: {volume:.2f} ml")
                
            except Exception as e:
                print(f"Error processing {view} {phase}: {str(e)}")
                continue
    
    # Calculate EF if we have all required data
    if (volumes['4CH']['ED'] and volumes['2CH']['ED'] and 
        volumes['4CH']['ES'] and volumes['2CH']['ES']):
        
        ed_vol = (volumes['4CH']['ED']['volume'] + volumes['2CH']['ED']['volume']) / 2
        es_vol = (volumes['4CH']['ES']['volume'] + volumes['2CH']['ES']['volume']) / 2
        sv = ed_vol - es_vol
        ef = (sv / ed_vol) * 100 if ed_vol > 0 else 0
        
        # Visualization of EF calculation
        print("\n6. Calculating Ejection Fraction...")
        plt.figure(figsize=(10, 6))
        bars = plt.bar(['EDV', 'ESV', 'SV'], [ed_vol, es_vol, sv], color=['blue', 'red', 'green'])
        plt.ylabel('Volume (ml)')
        plt.title(f'Ejection Fraction Calculation\nEF = (SV/EDV)  100 = ({sv:.1f}/{ed_vol:.1f})  100 = {ef:.1f}%')
        
        # Add value labels
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.1f} ml',
                    ha='center', va='bottom')
        
        plt.grid(alpha=0.3)
        plt.tight_layout()
        plt.show()
        
        return {
            'patient_id': patient_id,
            'EDV_4CH': volumes['4CH']['ED']['volume'],
            'ESV_4CH': volumes['4CH']['ES']['volume'],
            'EDV_2CH': volumes['2CH']['ED']['volume'],
            'ESV_2CH': volumes['2CH']['ES']['volume'],
            'EDV': ed_vol,
            'ESV': es_vol,
            'SV': sv,
            'EF': ef
        }
    else:
        print("\nIncomplete data for EF calculation")
        missing = []
        if not volumes['4CH']['ED']: missing.append("4CH ED")
        if not volumes['2CH']['ED']: missing.append("2CH ED")
        if not volumes['4CH']['ES']: missing.append("4CH ES")
        if not volumes['2CH']['ES']: missing.append("2CH ES")
        print(f"Missing: {', '.join(missing)}")
        return None

# Main processing - process one patient at a time
base_dir = r"E:\echo heart\Resources"
patient_dirs = [os.path.join(base_dir, d) for d in os.listdir(base_dir) 
               if os.path.isdir(os.path.join(base_dir, d))]

for patient_dir in patient_dirs[:1]:  # Just process first patient for demonstration
    result = process_patient_with_visualization(patient_dir)
    
    if result:
        print("\nFinal Results:")
        print(f"Patient: {result['patient_id']}")
        print(f"4CH ED Volume: {result['EDV_4CH']:.2f} ml")
        print(f"4CH ES Volume: {result['ESV_4CH']:.2f} ml")
        print(f"2CH ED Volume: {result['EDV_2CH']:.2f} ml")
        print(f"2CH ES Volume: {result['ESV_2CH']:.2f} ml")
        print(f"Average ED Volume: {result['EDV']:.2f} ml")
        print(f"Average ES Volume: {result['ESV']:.2f} ml")
        print(f"Stroke Volume: {result['SV']:.2f} ml")
        print(f"Ejection Fraction: {result['EF']:.2f} %")
    else:
        print("\nCould not calculate complete results for this patient")
```

```python
import os
import numpy as np
import nibabel as nib
import cv2
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
import math

def rotate_image_clockwise(image, angle=90):
    """Rotate image clockwise by specified angle (degrees) with visualization"""
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, -angle, 1.0)  # Negative angle for clockwise
    rotated = cv2.warpAffine(image, M, (w, h))
    
    # Visualization
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(image, cmap='gray')
    plt.title("Original Mask")
    plt.axis('off')
    
    plt.subplot(122)
    plt.imshow(rotated, cmap='gray')
    plt.title(f"Rotated Clockwise ({angle})")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    return rotated

def visualize_mask_creation(gt_data):
    """Visualize the mask creation process"""
    mask = np.zeros_like(gt_data, dtype=np.uint8)
    mask[gt_data == 1] = 255
    
    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.imshow(gt_data, cmap='gray')
    plt.title("Original Ground Truth")
    plt.axis('off')
    
    plt.subplot(122)
    plt.imshow(mask, cmap='gray')
    plt.title("LV Mask (White=LV, Black=Background)")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    return mask

def detect_landmarks_with_visualization(seg_binary):
    """Detect landmarks with visualization of each step"""
    seg_binary = (seg_binary > 0).astype(np.uint8) * 255
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    
    if not contours:
        raise ValueError("No LV contour found")
    
    # Create visualization canvas
    vis = np.stack([seg_binary]*3, axis=-1)
    
    # Get largest contour
    contour = max(contours, key=cv2.contourArea).squeeze()
    x = contour[:, 0].astype(float)
    y = contour[:, 1].astype(float)
    
    # Draw contour
    for i in range(len(contour)):
        cv2.circle(vis, tuple(contour[i][0]), 1, (0, 255, 0), -1)
    
    # Smooth contour
    x_smooth = gaussian_filter1d(x, sigma=5, mode='wrap')
    y_smooth = gaussian_filter1d(y, sigma=5, mode='wrap')
    smooth_contour = np.column_stack([x_smooth, y_smooth]).astype(int)
    
    # Compute curvature
    dx = np.gradient(x_smooth)
    dy = np.gradient(y_smooth)
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(dx * ddy - dy * ddx) / (dx**2 + dy**2 + 1e-10)**1.5
    curvature = (curvature - np.min(curvature)) / (np.max(curvature) - np.min(curvature) + 1e-10)
    
    # Find apex (min y)
    apex_idx = np.argmin(y)
    apex_point = (int(x[apex_idx]), int(y[apex_idx]))
    cv2.circle(vis, apex_point, 5, (255, 0, 0), -1)  # Red
    
    # Find peaks in curvature
    peaks, _ = find_peaks(curvature, prominence=0.0005)
    
    # Visualize curvature
    plt.figure(figsize=(15, 5))
    plt.subplot(131)
    plt.plot(curvature)
    plt.title("Curvature Along Contour")
    plt.xlabel("Contour point index")
    plt.ylabel("Normalized curvature")
    
    plt.subplot(132)
    plt.plot(x, y, 'g-', label='Original')
    plt.plot(x_smooth, y_smooth, 'b-', label='Smoothed')
    plt.plot(x[peaks], y[peaks], 'ro', label='Peaks')
    plt.plot(apex_point[0], apex_point[1], 'mo', markersize=10, label='Apex')
    plt.title("Contour with Peaks")
    plt.legend()
    plt.gca().invert_yaxis()
    
    plt.subplot(133)
    plt.imshow(vis)
    plt.title("Detected Contour Points (Green)")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    return apex_point, peaks, x, y, curvature

def compute_disk_diameters_with_visualization(seg_binary, apex, peaks, x, y, pixel_spacing, view, phase):
    """Compute disk diameters with full visualization"""
    # Create filled mask
    seg_binary = (seg_binary > 0).astype(np.uint8)
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    max_contour = max(contours, key=cv2.contourArea)
    filled_mask = np.zeros_like(seg_binary)
    cv2.drawContours(filled_mask, [max_contour], -1, 1, thickness=cv2.FILLED)
    
    # Create boundary
    kernel = np.ones((3,3), np.uint8)
    dilation = cv2.dilate(filled_mask, kernel)
    erosion = cv2.erode(filled_mask, kernel)
    boundary = (dilation - erosion > 0).astype(np.uint8)
    
    # Find septal and lateral points
    max_y = max(y)
    min_y = min(y)
    basal_threshold = max_y - 0.1 * (max_y - min_y)
    basal_peaks = [i for i in peaks if y[i] >= basal_threshold]
    
    if len(basal_peaks) >= 2:
        septal_idx = basal_peaks[0]
        lateral_idx = basal_peaks[1]
    else:
        mean_y = np.mean(y)
        lower_indices = [i for i in range(len(y)) if y[i] > mean_y]
        septal_idx = min(lower_indices, key=lambda i: x[i])
        lateral_idx = max(lower_indices, key=lambda i: x[i])
    
    septal_point = (int(x[septal_idx]), int(y[septal_idx]))
    lateral_point = (int(x[lateral_idx]), int(y[lateral_idx]))
    mitral_mid = ((septal_point[0] + lateral_point[0]) // 2, (septal_point[1] + lateral_point[1]) // 2)
    
    # Long axis
    dx = apex[0] - mitral_mid[0]
    dy = apex[1] - mitral_mid[1]
    L_px = np.sqrt(dx**2 + dy**2)
    L_cm = L_px * pixel_spacing[0]
    
    # Perpendicular vector
    perp_vector = np.array([-dy, dx]) / L_px if L_px > 0 else np.array([1, 0])
    
    # Visualization setup
    vis = np.stack([seg_binary*255]*3, axis=-1)
    cv2.line(vis, apex, mitral_mid, (0, 255, 255), 2)  # Yellow long axis
    cv2.circle(vis, apex, 5, (255, 0, 0), -1)  # Red apex
    cv2.circle(vis, septal_point, 5, (0, 0, 255), -1)  # Blue septal
    cv2.circle(vis, lateral_point, 5, (0, 255, 0), -1)  # Green lateral
    cv2.circle(vis, mitral_mid, 5, (255, 255, 0), -1)  # Cyan mitral mid
    
    nr_disks = 20
    diameters_cm = []
    disk_lines = []
    
    for i in range(nr_disks):
        t = (i + 0.5) / nr_disks
        center_x = mitral_mid[0] + t * dx
        center_y = mitral_mid[1] + t * dy
        
        left_point = None
        right_point = None
        
        # Search left
        for dist in range(1, 150):
            px = int(round(center_x - dist * perp_vector[0]))
            py = int(round(center_y - dist * perp_vector[1]))
            if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                break
            if boundary[py, px]:
                left_point = (px, py)
                break
        
        # Search right
        for dist in range(1, 150):
            px = int(round(center_x + dist * perp_vector[0]))
            py = int(round(center_y + dist * perp_vector[1]))
            if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                break
            if boundary[py, px]:
                right_point = (px, py)
                break
        
        if left_point and right_point:
            dx_px = right_point[0] - left_point[0]
            dy_px = right_point[1] - left_point[1]
            diameter_cm = np.sqrt((dx_px * pixel_spacing[0])**2 + (dy_px * pixel_spacing[1])**2)
            diameters_cm.append(diameter_cm)
            
            # Add to visualization
            cv2.line(vis, left_point, right_point, (255, 0, 255), 1)  # Purple disk line
            disk_lines.append((left_point, right_point))
    
    # Final visualization
    plt.figure(figsize=(15, 5))
    plt.subplot(131)
    plt.imshow(seg_binary, cmap='gray')
    plt.title(f"{view} {phase} - Binary Mask")
    plt.axis('off')
    
    plt.subplot(132)
    plt.imshow(boundary, cmap='gray')
    plt.title("Boundary Detection")
    plt.axis('off')
    
    plt.subplot(133)
    plt.imshow(vis)
    plt.title(f"Landmarks and Disks\nApex(Red), Septal(Blue), Lateral(Green)\nLong Axis: {L_cm:.2f} cm")
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    # Calculate dead space area
    dead_area_cm2 = 0.0
    if disk_lines:
        polygon_points = []
        for left, right in disk_lines:
            polygon_points.append(left)
        for left, right in reversed(disk_lines):
            polygon_points.append(right)
        
        if len(polygon_points) > 2:
            disk_area_mask = np.zeros_like(filled_mask, dtype=np.uint8)
            pts = np.array([polygon_points], dtype=np.int32)
            cv2.fillPoly(disk_area_mask, [pts], 1)
            dead_space_mask = filled_mask - disk_area_mask
            dead_space_mask = np.clip(dead_space_mask, 0, 1)
            dead_area_px = np.sum(dead_space_mask)
            dead_area_cm2 = dead_area_px * pixel_spacing[0] * pixel_spacing[1]
            
            # Visualize dead space
            plt.figure(figsize=(10, 5))
            plt.subplot(121)
            plt.imshow(disk_area_mask, cmap='gray')
            plt.title("Disk Area Coverage")
            plt.axis('off')
            
            plt.subplot(122)
            plt.imshow(dead_space_mask, cmap='gray')
            plt.title(f"Dead Space Area: {dead_area_cm2:.2f} cm")
            plt.axis('off')
            plt.tight_layout()
            plt.show()
    
    return L_cm, diameters_cm, dead_area_cm2

def calculate_simpson_volume(L_cm, diameters_cm, dead_area_cm2, nr_disks=20):
    """Calculate volume using Simpson's method with visualization"""
    h_cm = L_cm / nr_disks
    disk_volumes = [(math.pi/4) * d**2 * h_cm for d in diameters_cm]
    total_disk_volume = sum(disk_volumes)
    dead_volume = dead_area_cm2 * h_cm
    total_volume = total_disk_volume + dead_volume
    
    # Visualization
    plt.figure(figsize=(12, 6))
    plt.bar(range(len(disk_volumes)), disk_volumes, color='skyblue', label='Disk Volumes')
    plt.axhline(y=total_disk_volume, color='red', linestyle='--', label=f'Total Disk Volume: {total_disk_volume:.2f} ml')
    plt.axhline(y=dead_volume, color='green', linestyle='--', label=f'Dead Space Volume: {dead_volume:.2f} ml')
    plt.axhline(y=total_volume, color='purple', linestyle='-', label=f'Total Volume: {total_volume:.2f} ml')
    plt.xlabel("Disk Number")
    plt.ylabel("Volume (ml)")
    plt.title("Simpson's Method Volume Calculation")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return total_volume

def process_patient_with_visualization(patient_dir):
    """Process one patient with full visualization of each step"""
    patient_id = os.path.basename(patient_dir)
    print(f"\n{'='*50}")
    print(f"Processing Patient: {patient_id}")
    print(f"{'='*50}\n")
    
    volumes = {'2CH': {'ED': None, 'ES': None}, '4CH': {'ED': None, 'ES': None}}
    
    for file in os.listdir(patient_dir):
        # Handle both .nii.gz and .niigz extensions
        if file.endswith('_gt.niigz') or file.endswith('_gt.nii.gz'):
            # Determine view and phase
            view = '4CH' if '4ch' in file.lower() else '2CH' if '2ch' in file.lower() else None
            phase = 'ED' if 'ed' in file.lower() else 'ES' if 'es' in file.lower() else None
            
            if not view or not phase:
                continue
            
            print(f"\nProcessing {view} {phase} view...")
            
            # Load data
            gt_path = os.path.join(patient_dir, file)
            
            # Handle both compressed formats
            if file.endswith('.niigz'):
                # Create temporary decompressed file
                import tempfile
                import gzip
                import shutil
                
                with tempfile.NamedTemporaryFile(suffix='.nii.gz', delete=False) as tmp_file:
                    with gzip.open(gt_path, 'rb') as f_in:
                        shutil.copyfileobj(f_in, tmp_file)
                    gt_path = tmp_file.name
                    gt_img = nib.load(gt_path)
                    os.unlink(tmp_file.name)  # Clean up temporary file
            else:
                gt_img = nib.load(gt_path)
            
            gt_data = gt_img.get_fdata()
            pixel_spacing = gt_img.header.get_zooms()[:2]
            
            # Step 1: Mask creation visualization
            print("\n1. Creating LV Mask...")
            mask = visualize_mask_creation(gt_data)
            
            # Step 2: Rotation visualization (clockwise)
            print("\n2. Rotating Image Clockwise...")
            rotated_mask = rotate_image_clockwise(mask, 90)
            
            # Step 3: Landmark detection visualization
            print("\n3. Detecting Landmarks...")
            try:
                apex_point, peaks, x, y, curvature = detect_landmarks_with_visualization(rotated_mask)
                
                # Step 4: Disk diameter calculation visualization
                print("\n4. Calculating Disk Diameters...")
                L_cm, diameters_cm, dead_area_cm2 = compute_disk_diameters_with_visualization(
                    rotated_mask, apex_point, peaks, x, y, pixel_spacing, view, phase
                )
                
                # Step 5: Volume calculation visualization
                print("\n5. Calculating Volume...")
                volume = calculate_simpson_volume(L_cm, diameters_cm, dead_area_cm2)
                
                volumes[view][phase] = {
                    'L_cm': L_cm,
                    'diameters': diameters_cm,
                    'dead_area': dead_area_cm2,
                    'volume': volume
                }
                
                print(f"\n{view} {phase} Volume: {volume:.2f} ml")
                
            except Exception as e:
                print(f"Error processing {view} {phase}: {str(e)}")
                continue
    
    # Calculate EF if we have all required data
    if (volumes['4CH']['ED'] and volumes['2CH']['ED'] and 
        volumes['4CH']['ES'] and volumes['2CH']['ES']):
        
        ed_vol = (volumes['4CH']['ED']['volume'] + volumes['2CH']['ED']['volume']) / 2
        es_vol = (volumes['4CH']['ES']['volume'] + volumes['2CH']['ES']['volume']) / 2
        sv = ed_vol - es_vol
        ef = (sv / ed_vol) * 100 if ed_vol > 0 else 0
        
        # Visualization of EF calculation
        print("\n6. Calculating Ejection Fraction...")
        plt.figure(figsize=(10, 6))
        bars = plt.bar(['EDV', 'ESV', 'SV'], [ed_vol, es_vol, sv], color=['blue', 'red', 'green'])
        plt.ylabel('Volume (ml)')
        plt.title(f'Ejection Fraction Calculation\nEF = (SV/EDV)  100 = ({sv:.1f}/{ed_vol:.1f})  100 = {ef:.1f}%')
        
        # Add value labels
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.1f} ml',
                    ha='center', va='bottom')
        
        plt.grid(alpha=0.3)
        plt.tight_layout()
        plt.show()
        
        return {
            'patient_id': patient_id,
            'EDV_4CH': volumes['4CH']['ED']['volume'],
            'ESV_4CH': volumes['4CH']['ES']['volume'],
            'EDV_2CH': volumes['2CH']['ED']['volume'],
            'ESV_2CH': volumes['2CH']['ES']['volume'],
            'EDV': ed_vol,
            'ESV': es_vol,
            'SV': sv,
            'EF': ef
        }
    else:
        print("\nIncomplete data for EF calculation")
        missing = []
        if not volumes['4CH']['ED']: missing.append("4CH ED")
        if not volumes['2CH']['ED']: missing.append("2CH ED")
        if not volumes['4CH']['ES']: missing.append("4CH ES")
        if not volumes['2CH']['ES']: missing.append("2CH ES")
        print(f"Missing: {', '.join(missing)}")
        return None

# Main processing - process one patient at a time
base_dir = r"E:\echo heart\Resources"
patient_dirs = [os.path.join(base_dir, d) for d in os.listdir(base_dir) 
               if os.path.isdir(os.path.join(base_dir, d))]

for patient_dir in patient_dirs[:1]:  # Just process first patient for demonstration
    result = process_patient_with_visualization(patient_dir)
    
    if result:
        print("\nFinal Results:")
        print(f"Patient: {result['patient_id']}")
        print(f"4CH ED Volume: {result['EDV_4CH']:.2f} ml")
        print(f"4CH ES Volume: {result['ESV_4CH']:.2f} ml")
        print(f"2CH ED Volume: {result['EDV_2CH']:.2f} ml")
        print(f"2CH ES Volume: {result['ESV_2CH']:.2f} ml")
        print(f"Average ED Volume: {result['EDV']:.2f} ml")
        print(f"Average ES Volume: {result['ESV']:.2f} ml")
        print(f"Stroke Volume: {result['SV']:.2f} ml")
        print(f"Ejection Fraction: {result['EF']:.2f} %")
    else:
        print("\nCould not calculate complete results for this patient")
```

```python
import os
import numpy as np
import nibabel as nib
import cv2
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
import math
import gzip
import shutil
import tempfile
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

def handle_compressed_file(filepath):
    """Handle both .nii.gz and .niigz files"""
    if filepath.endswith('.niigz'):
        with tempfile.NamedTemporaryFile(suffix='.nii.gz', delete=False) as tmp_file:
            with gzip.open(filepath, 'rb') as f_in:
                shutil.copyfileobj(f_in, tmp_file)
            return tmp_file.name
    return filepath

def rotate_image_clockwise(image, angle=90):
    """Rotate image clockwise with better boundary handling"""
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, -angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_NEAREST)
    return rotated

def detect_landmarks(seg_binary, delta_y_fraction=0.1, y_tolerance=10):
    """Improved landmark detection with curvature-based refinement."""
    # Convert to binary and find contours
    seg_binary = (seg_binary > 0).astype(np.uint8) * 255
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        raise ValueError("No LV contour found")
    
    # Use the largest contour
    contour = max(contours, key=cv2.contourArea).squeeze()
    x = contour[:, 0].astype(float)
    y = contour[:, 1].astype(float)
    
    # Smooth contour
    x_smooth = gaussian_filter1d(x, sigma=5, mode='wrap')
    y_smooth = gaussian_filter1d(y, sigma=5, mode='wrap')
    
    # Compute curvature
    dx = np.gradient(x_smooth)
    dy = np.gradient(y_smooth)
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(dx * ddy - dy * ddx) / (dx**2 + dy**2 + 1e-10)**1.5
    curvature = (curvature - np.min(curvature)) / (np.max(curvature) - np.min(curvature) + 1e-10)
    
    # Find apex: min y
    apex_idx = np.argmin(y)
    apex_point = (int(x[apex_idx]), int(y[apex_idx]))
    
    # Find all peaks in curvature
    peaks, _ = find_peaks(curvature, prominence=0.0005)
    
    # Define basal region: y > max_y * (1 - delta_y_fraction)
    max_y = max(y)
    min_y = min(y)
    basal_threshold = max_y - delta_y_fraction * (max_y - min_y)
    basal_peaks = [i for i in peaks if y[i] >= basal_threshold]
    
    if len(basal_peaks) >= 2:
        # Sort peaks by curvature (highest to lowest)
        sorted_peaks = sorted(basal_peaks, key=lambda i: curvature[i], reverse=True)
        # Assign first highest peak as septal, second as lateral
        septal_idx = sorted_peaks[0]
        lateral_idx = sorted_peaks[1]
        
        # Verify they are on one line (similar y-coordinates)
        if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
            # Find a lateral peak with similar y-coordinate to septal
            septal_y = y[septal_idx]
            lateral_candidates = [i for i in basal_peaks if abs(y[i] - septal_y) <= y_tolerance]
            if len(lateral_candidates) >= 2:
                # Select the highest curvature peak among candidates, excluding septal
                lateral_candidates = [i for i in lateral_candidates if i != septal_idx]
                if lateral_candidates:
                    lateral_idx = max(lateral_candidates, key=lambda i: curvature[i])
                else:
                    raise ValueError("No suitable lateral peak found with similar y-coordinate")
            else:
                raise ValueError("Not enough peaks with similar y-coordinates")
    else:
        # Fallback: points with y > mean(y), min and max x
        mean_y = np.mean(y)
        lower_indices = [i for i in range(len(y)) if y[i] > mean_y]
        if len(lower_indices) >= 2:
            septal_idx = min(lower_indices, key=lambda i: x[i])
            lateral_idx = max(lower_indices, key=lambda i: x[i])
            # Verify y-alignment in fallback
            if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
                raise ValueError("Fallback septal and lateral points not on one line")
        else:
            raise ValueError("Not enough points for fallback")
    
    septal_point = (int(x[septal_idx]), int(y[septal_idx]))
    lateral_point = (int(x[lateral_idx]), int(y[lateral_idx]))
    
    # Mitral mid: midpoint between septal and lateral
    mitral_mid = ((septal_point[0] + lateral_point[0]) / 2, (septal_point[1] + lateral_point[1]) / 2)
    
    return apex_point, septal_point, lateral_point, mitral_mid

def compute_disk_diameters(seg_binary, apex, mitral_mid, pixel_spacing=(0.1, 0.1), nr_disks=20):
    """Compute disk diameters, long axis length, and dead space for a single view."""
    # Precompute filled LV mask
    contours, _ = cv2.findContours(seg_binary.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        return 0.0, [], [], 0.0
    
    max_contour = max(contours, key=cv2.contourArea)
    filled_mask = np.zeros_like(seg_binary)
    cv2.drawContours(filled_mask, [max_contour], -1, 1, thickness=cv2.FILLED)
    
    # Compute boundary
    kernel = np.ones((3,3), np.uint8)
    dilation = cv2.dilate(filled_mask, kernel)
    erosion = cv2.erode(filled_mask, kernel)
    boundary = (dilation - erosion > 0).astype(np.uint8)
    
    # Long axis vector
    dx = apex[0] - mitral_mid[0]
    dy = apex[1] - mitral_mid[1]
    L_px = np.sqrt(dx**2 + dy**2)
    L_cm = L_px * pixel_spacing[0]  # Assume isotropic spacing
    
    if L_cm == 0:
        return L_cm, [], [], 0.0
    
    # Perpendicular vector
    perp_vector = np.array([-dy, dx]) / L_px if L_px > 0 else np.array([1, 0])
    
    diameters_cm = []
    disk_info = []
    
    for i in range(nr_disks):
        t = (i + 0.5) / nr_disks
        center_x = mitral_mid[0] + t * dx
        center_y = mitral_mid[1] + t * dy
        
        left_point = None
        right_point = None
        
        # Search in both directions
        for direction in [-1, 1]:
            for dist in range(1, 150):
                px = int(round(center_x + direction * dist * perp_vector[0]))
                py = int(round(center_y + direction * dist * perp_vector[1]))
                
                if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                    break
                
                if boundary[py, px]:
                    if direction == -1:
                        left_point = (px, py)
                    else:
                        right_point = (px, py)
                    break
        
        if left_point and right_point:
            dx_px = right_point[0] - left_point[0]
            dy_px = right_point[1] - left_point[1]
            diameter_cm = np.sqrt((dx_px * pixel_spacing[0])**2 + (dy_px * pixel_spacing[1])**2)
            diameters_cm.append(diameter_cm)
            disk_info.append({
                'center': (center_x, center_y),
                'left': left_point,
                'right': right_point,
                'diameter_cm': diameter_cm
            })
        else:
            diameters_cm.append(0.0)
            disk_info.append({
                'center': (center_x, center_y),
                'left': None,
                'right': None,
                'diameter_cm': 0.0
            })
    
    # Calculate dead space area
    dead_area_cm2 = 0.0
    
    if len([d for d in disk_info if d['left'] and d['right']]) >= 2:
        # Create polygon from disk endpoints
        polygon_points = []
        
        # Add left points from base to apex
        for disk in disk_info:
            if disk['left']:
                polygon_points.append(disk['left'])
        
        # Add right points from apex to base
        for disk in reversed(disk_info):
            if disk['right']:
                polygon_points.append(disk['right'])
        
        # Create mask for disk-covered area
        disk_area_mask = np.zeros_like(filled_mask, dtype=np.uint8)
        if len(polygon_points) > 2:
            pts = np.array([polygon_points], dtype=np.int32)
            cv2.fillPoly(disk_area_mask, [pts], 1)
            
            # Calculate dead space as LV area not covered by disks
            dead_space_mask = filled_mask - disk_area_mask
            dead_space_mask = np.clip(dead_space_mask, 0, 1)  # Ensure binary
            
            # Convert to cm
            dead_area_px = np.sum(dead_space_mask)
            dead_area_cm2 = dead_area_px * pixel_spacing[0] * pixel_spacing[1]
    
    return L_cm, diameters_cm, disk_info, dead_area_cm2

def compute_biplane_volume(L4_cm, L2_cm, diameters4_cm, diameters2_cm, dead_area4_cm2, dead_area2_cm2, nr_disks=20):
    """Compute biplane Simpson's volume with dead space correction."""
    # Use average long axis length
    L_avg_cm = (L4_cm + L2_cm) / 2
    h_cm = L_avg_cm / nr_disks
    
    # Simpson's volume from disks
    total_volume = 0.0
    disk_volumes = []
    
    for i in range(nr_disks):
        a_i = diameters4_cm[i] if i < len(diameters4_cm) else 0
        b_i = diameters2_cm[i] if i < len(diameters2_cm) else 0
        disk_volume = (np.pi / 4) * a_i * b_i * h_cm
        total_volume += disk_volume
        disk_volumes.append(disk_volume)
    
    # Calculate dead volume
    avg_dead_area_cm2 = (dead_area4_cm2 + dead_area2_cm2) / 2
    dead_volume = avg_dead_area_cm2 * h_cm
    
    # Add dead volume to total
    total_volume_with_dead = total_volume + dead_volume
    
    return total_volume_with_dead

def visualize_processing(seg_data, landmarks, disk_info, view, phase, pixel_spacing):
    """Visualize the processing steps for one view/phase"""
    apex, septal, lateral, mitral_mid = landmarks
    
    # Create figure
    plt.figure(figsize=(15, 5))
    
    # Original image with landmarks
    plt.subplot(131)
    plt.imshow(seg_data, cmap='gray')
    plt.scatter(apex[0], apex[1], c='red', s=80, marker='x', label='Apex')
    plt.scatter(septal[0], septal[1], c='blue', s=80, marker='x', label='Septal')
    plt.scatter(lateral[0], lateral[1], c='green', s=80, marker='x', label='Lateral')
    plt.scatter(mitral_mid[0], mitral_mid[1], c='magenta', s=100, marker='+', label='Mitral Mid')
    plt.plot([apex[0], mitral_mid[0]], [apex[1], mitral_mid[1]], 'y--', label='Long Axis')
    plt.plot([septal[0], lateral[0]], [septal[1], lateral[1]], 'c--', label='Mitral Line')
    plt.title(f'{view} {phase} - Landmarks')
    plt.legend()
    
    # Measurement disks
    plt.subplot(132)
    plt.imshow(seg_data, cmap='gray')
    for disk in disk_info:
        if disk['left'] and disk['right']:
            plt.plot([disk['left'][0], disk['right'][0]], [disk['left'][1], disk['right'][1]], 'r-', linewidth=2)
        plt.scatter(disk['center'][0], disk['center'][1], c='yellow', s=20)
    plt.title(f'{view} {phase} - Measurement Disks')
    
    # Volume contribution per disk
    plt.subplot(133)
    if disk_info and 'diameter_cm' in disk_info[0]:
        diameters = [d['diameter_cm'] for d in disk_info]
        plt.plot(diameters, 'bo-')
        plt.title(f'{view} {phase} - Disk Diameters')
        plt.xlabel('Disk Number')
        plt.ylabel('Diameter (cm)')
        plt.grid(True)
    
    plt.tight_layout()
    plt.show()

def process_patient(patient_dir, visualize=False):
    """Process one patient with improved volume calculation"""
    patient_id = os.path.basename(patient_dir)
    print(f"\nProcessing Patient: {patient_id}")
    
    volumes = {'2CH': {'ED': None, 'ES': None}, '4CH': {'ED': None, 'ES': None}}
    all_landmarks = {}
    all_disk_info = {}
    
    for file in os.listdir(patient_dir):
        if not (file.endswith('_gt.niigz') or file.endswith('_gt.nii.gz')):
            continue
            
        try:
            # Determine view and phase
            view = '4CH' if '4ch' in file.lower() else '2CH' if '2ch' in file.lower() else None
            phase = 'ED' if 'ed' in file.lower() else 'ES' if 'es' in file.lower() else None
            
            if not view or not phase:
                continue
                
            print(f"\nProcessing {view} {phase}...")
            
            # Handle compressed files
            filepath = os.path.join(patient_dir, file)
            temp_file = None
            
            if file.endswith('.niigz'):
                temp_file = handle_compressed_file(filepath)
                gt_img = nib.load(temp_file)
            else:
                gt_img = nib.load(filepath)
                
            gt_data = gt_img.get_fdata()
            pixel_spacing = gt_img.header.get_zooms()[:2]
            
            # Create mask and rotate
            mask = np.zeros_like(gt_data, dtype=np.uint8)
            mask[gt_data == 1] = 255
            rotated_mask = rotate_image_clockwise(mask)
            
            # Detect landmarks
            apex, septal, lateral, mitral_mid = detect_landmarks(rotated_mask)
            
            # Calculate disk diameters and dead space
            L_cm, diameters_cm, disk_info, dead_area_cm2 = compute_disk_diameters(
                rotated_mask, apex, mitral_mid, pixel_spacing
            )
            
            # Store results
            volumes[view][phase] = {
                'volume': None,  # Will be calculated after both views are processed
                'L_cm': L_cm,
                'diameters_cm': diameters_cm,
                'dead_area_cm2': dead_area_cm2,
                'status': 'success'
            }
            
            all_landmarks[(view, phase)] = (apex, septal, lateral, mitral_mid)
            all_disk_info[(view, phase)] = disk_info
            
            # Visualize if requested
            if visualize:
                visualize_processing(rotated_mask, (apex, septal, lateral, mitral_mid), 
                                  disk_info, view, phase, pixel_spacing)
            
        except Exception as e:
            print(f"Error processing {file}: {str(e)}")
            if view and phase:
                volumes[view][phase] = {
                    'volume': None,
                    'status': f'failed: {str(e)}'
                }
        finally:
            if temp_file and os.path.exists(temp_file):
                os.unlink(temp_file)
    
    # Calculate volumes and EF if we have all required data
    missing = []
    for view in ['2CH', '4CH']:
        for phase in ['ED', 'ES']:
            if not volumes[view][phase] or volumes[view][phase]['status'] != 'success':
                missing.append(f"{view} {phase}")
    
    if missing:
        print(f"\nIncomplete data for EF calculation. Missing: {', '.join(missing)}")
    else:
        # Calculate EDV and ESV using biplane Simpson's method
        try:
            # ED volumes
            ed_vol = compute_biplane_volume(
                volumes['4CH']['ED']['L_cm'],
                volumes['2CH']['ED']['L_cm'],
                volumes['4CH']['ED']['diameters_cm'],
                volumes['2CH']['ED']['diameters_cm'],
                volumes['4CH']['ED']['dead_area_cm2'],
                volumes['2CH']['ED']['dead_area_cm2']
            )
            
            # ES volumes
            es_vol = compute_biplane_volume(
                volumes['4CH']['ES']['L_cm'],
                volumes['2CH']['ES']['L_cm'],
                volumes['4CH']['ES']['diameters_cm'],
                volumes['2CH']['ES']['diameters_cm'],
                volumes['4CH']['ES']['dead_area_cm2'],
                volumes['2CH']['ES']['dead_area_cm2']
            )
            
            # Calculate EF
            ef = ((ed_vol - es_vol) / ed_vol) * 100 if ed_vol > 0 else 0
            
            # Update volumes dictionary
            volumes['EDV'] = ed_vol
            volumes['ESV'] = es_vol
            volumes['EF'] = ef
            
            print(f"\nEF Calculation Successful:")
            print(f"EDV: {ed_vol:.2f} ml")
            print(f"ESV: {es_vol:.2f} ml")
            print(f"EF: {ef:.2f}%")
            
            # Final visualization
            if visualize:
                plt.figure(figsize=(10, 5))
                bars = plt.bar(['EDV', 'ESV'], [ed_vol, es_vol], color=['blue', 'red'])
                plt.ylabel('Volume (ml)')
                plt.title(f'Ejection Fraction Calculation\nEF = {ef:.1f}%')
                for bar in bars:
                    height = bar.get_height()
                    plt.text(bar.get_x() + bar.get_width()/2., height,
                            f'{height:.1f} ml', ha='center', va='bottom')
                plt.show()
                
        except Exception as e:
            print(f"Error calculating volumes: {str(e)}")
    
    return volumes, all_landmarks, all_disk_info

# Process one patient
base_dir = r"E:\echo heart\Resources"
patient_dirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

if patient_dirs:
    patient_volumes, landmarks, disk_info = process_patient(
        os.path.join(base_dir, patient_dirs[0]), 
        visualize=True
    )
else:
    print("No patient directories found")
```

```python
import os
import numpy as np
import nibabel as nib
import cv2
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
import math
import gzip
import shutil
import tempfile
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

def handle_compressed_file(filepath):
    """Handle both .nii.gz and .niigz files"""
    if filepath.endswith('.niigz'):
        with tempfile.NamedTemporaryFile(suffix='.nii.gz', delete=False) as tmp_file:
            with gzip.open(filepath, 'rb') as f_in:
                shutil.copyfileobj(f_in, tmp_file)
            return tmp_file.name
    return filepath

def rotate_image_clockwise(image, angle=90):
    """Rotate image clockwise with better boundary handling"""
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, -angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_NEAREST)
    return rotated

def detect_landmarks(seg_binary, delta_y_fraction=0.1, y_tolerance=10):
    """Improved landmark detection with curvature-based refinement."""
    # Convert to binary and find contours
    seg_binary = (seg_binary > 0).astype(np.uint8) * 255
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        raise ValueError("No LV contour found")
    
    # Use the largest contour
    contour = max(contours, key=cv2.contourArea).squeeze()
    x = contour[:, 0].astype(float)
    y = contour[:, 1].astype(float)
    
    # Smooth contour
    x_smooth = gaussian_filter1d(x, sigma=5, mode='wrap')
    y_smooth = gaussian_filter1d(y, sigma=5, mode='wrap')
    
    # Compute curvature
    dx = np.gradient(x_smooth)
    dy = np.gradient(y_smooth)
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(dx * ddy - dy * ddx) / (dx**2 + dy**2 + 1e-10)**1.5
    curvature = (curvature - np.min(curvature)) / (np.max(curvature) - np.min(curvature) + 1e-10)
    
    # Find apex: min y
    apex_idx = np.argmin(y)
    apex_point = (int(x[apex_idx]), int(y[apex_idx]))
    
    # Find all peaks in curvature
    peaks, _ = find_peaks(curvature, prominence=0.0005)
    
    # Define basal region: y > max_y * (1 - delta_y_fraction)
    max_y = max(y)
    min_y = min(y)
    basal_threshold = max_y - delta_y_fraction * (max_y - min_y)
    basal_peaks = [i for i in peaks if y[i] >= basal_threshold]
    
    if len(basal_peaks) >= 2:
        # Sort peaks by curvature (highest to lowest)
        sorted_peaks = sorted(basal_peaks, key=lambda i: curvature[i], reverse=True)
        # Assign first highest peak as septal, second as lateral
        septal_idx = sorted_peaks[0]
        lateral_idx = sorted_peaks[1]
        
        # Verify they are on one line (similar y-coordinates)
        if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
            # Find a lateral peak with similar y-coordinate to septal
            septal_y = y[septal_idx]
            lateral_candidates = [i for i in basal_peaks if abs(y[i] - septal_y) <= y_tolerance]
            if len(lateral_candidates) >= 2:
                # Select the highest curvature peak among candidates, excluding septal
                lateral_candidates = [i for i in lateral_candidates if i != septal_idx]
                if lateral_candidates:
                    lateral_idx = max(lateral_candidates, key=lambda i: curvature[i])
                else:
                    raise ValueError("No suitable lateral peak found with similar y-coordinate")
            else:
                raise ValueError("Not enough peaks with similar y-coordinates")
    else:
        # Fallback: points with y > mean(y), min and max x
        mean_y = np.mean(y)
        lower_indices = [i for i in range(len(y)) if y[i] > mean_y]
        if len(lower_indices) >= 2:
            septal_idx = min(lower_indices, key=lambda i: x[i])
            lateral_idx = max(lower_indices, key=lambda i: x[i])
            # Verify y-alignment in fallback
            if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
                raise ValueError("Fallback septal and lateral points not on one line")
        else:
            raise ValueError("Not enough points for fallback")
    
    septal_point = (int(x[septal_idx]), int(y[septal_idx]))
    lateral_point = (int(x[lateral_idx]), int(y[lateral_idx]))
    
    # Mitral mid: midpoint between septal and lateral
    mitral_mid = ((septal_point[0] + lateral_point[0]) / 2, (septal_point[1] + lateral_point[1]) / 2)
    
    return apex_point, septal_point, lateral_point, mitral_mid

def compute_disk_diameters(seg_binary, apex, mitral_mid, pixel_spacing=(0.1, 0.1), nr_disks=20):
    """Compute disk diameters, long axis length, and dead space for a single view."""
    # Precompute filled LV mask
    contours, _ = cv2.findContours(seg_binary.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        return 0.0, [], [], 0.0
    
    max_contour = max(contours, key=cv2.contourArea)
    filled_mask = np.zeros_like(seg_binary)
    cv2.drawContours(filled_mask, [max_contour], -1, 1, thickness=cv2.FILLED)
    
    # Compute boundary
    kernel = np.ones((3,3), np.uint8)
    dilation = cv2.dilate(filled_mask, kernel)
    erosion = cv2.erode(filled_mask, kernel)
    boundary = (dilation - erosion > 0).astype(np.uint8)
    
    # Long axis vector
    dx = apex[0] - mitral_mid[0]
    dy = apex[1] - mitral_mid[1]
    L_px = np.sqrt(dx**2 + dy**2)
    L_cm = L_px * pixel_spacing[0]  # Assume isotropic spacing
    
    if L_cm == 0:
        return L_cm, [], [], 0.0
    
    # Perpendicular vector
    perp_vector = np.array([-dy, dx]) / L_px if L_px > 0 else np.array([1, 0])
    
    diameters_cm = []
    disk_info = []
    
    for i in range(nr_disks):
        t = (i + 0.5) / nr_disks
        center_x = mitral_mid[0] + t * dx
        center_y = mitral_mid[1] + t * dy
        
        left_point = None
        right_point = None
        
        # Search in both directions
        for direction in [-1, 1]:
            for dist in range(1, 150):
                px = int(round(center_x + direction * dist * perp_vector[0]))
                py = int(round(center_y + direction * dist * perp_vector[1]))
                
                if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                    break
                
                if boundary[py, px]:
                    if direction == -1:
                        left_point = (px, py)
                    else:
                        right_point = (px, py)
                    break
        
        if left_point and right_point:
            dx_px = right_point[0] - left_point[0]
            dy_px = right_point[1] - left_point[1]
            diameter_cm = np.sqrt((dx_px * pixel_spacing[0])**2 + (dy_px * pixel_spacing[1])**2)
            diameters_cm.append(diameter_cm)
            disk_info.append({
                'center': (center_x, center_y),
                'left': left_point,
                'right': right_point,
                'diameter_cm': diameter_cm
            })
        else:
            diameters_cm.append(0.0)
            disk_info.append({
                'center': (center_x, center_y),
                'left': None,
                'right': None,
                'diameter_cm': 0.0
            })
    
    # Calculate dead space area
    dead_area_cm2 = 0.0
    
    if len([d for d in disk_info if d['left'] and d['right']]) >= 2:
        # Create polygon from disk endpoints
        polygon_points = []
        
        # Add left points from base to apex
        for disk in disk_info:
            if disk['left']:
                polygon_points.append(disk['left'])
        
        # Add right points from apex to base
        for disk in reversed(disk_info):
            if disk['right']:
                polygon_points.append(disk['right'])
        
        # Create mask for disk-covered area
        disk_area_mask = np.zeros_like(filled_mask, dtype=np.uint8)
        if len(polygon_points) > 2:
            pts = np.array([polygon_points], dtype=np.int32)
            cv2.fillPoly(disk_area_mask, [pts], 1)
            
            # Calculate dead space as LV area not covered by disks
            dead_space_mask = filled_mask - disk_area_mask
            dead_space_mask = np.clip(dead_space_mask, 0, 1)  # Ensure binary
            
            # Convert to cm
            dead_area_px = np.sum(dead_space_mask)
            dead_area_cm2 = dead_area_px * pixel_spacing[0] * pixel_spacing[1]
    
    return L_cm, diameters_cm, disk_info, dead_area_cm2

def compute_biplane_volume(L4_cm, L2_cm, diameters4_cm, diameters2_cm, dead_area4_cm2, dead_area2_cm2, nr_disks=20):
    """Compute biplane Simpson's volume with dead space correction."""
    # Use average long axis length
    L_avg_cm = (L4_cm + L2_cm) / 2
    h_cm = L_avg_cm / nr_disks
    
    # Simpson's volume from disks
    total_volume = 0.0
    disk_volumes = []
    
    for i in range(nr_disks):
        a_i = diameters4_cm[i] if i < len(diameters4_cm) else 0
        b_i = diameters2_cm[i] if i < len(diameters2_cm) else 0
        disk_volume = (np.pi / 4) * a_i * b_i * h_cm
        total_volume += disk_volume
        disk_volumes.append(disk_volume)
    
    # Calculate dead volume
    avg_dead_area_cm2 = (dead_area4_cm2 + dead_area2_cm2) / 2
    dead_volume = avg_dead_area_cm2 * h_cm
    
    # Add dead volume to total
    total_volume_with_dead = total_volume + dead_volume
    
    return total_volume_with_dead

def visualize_processing(seg_data, landmarks, disk_info, view, phase, pixel_spacing):
    """Visualize the processing steps for one view/phase"""
    apex, septal, lateral, mitral_mid = landmarks
    
    # Create figure
    plt.figure(figsize=(15, 5))
    
    # Original image with landmarks
    plt.subplot(131)
    plt.imshow(seg_data, cmap='gray')
    plt.scatter(apex[0], apex[1], c='red', s=80, marker='x', label='Apex')
    plt.scatter(septal[0], septal[1], c='blue', s=80, marker='x', label='Septal')
    plt.scatter(lateral[0], lateral[1], c='green', s=80, marker='x', label='Lateral')
    plt.scatter(mitral_mid[0], mitral_mid[1], c='magenta', s=100, marker='+', label='Mitral Mid')
    plt.plot([apex[0], mitral_mid[0]], [apex[1], mitral_mid[1]], 'y--', label='Long Axis')
    plt.plot([septal[0], lateral[0]], [septal[1], lateral[1]], 'c--', label='Mitral Line')
    plt.title(f'{view} {phase} - Landmarks')
    plt.legend()
    
    # Measurement disks
    plt.subplot(132)
    plt.imshow(seg_data, cmap='gray')
    for disk in disk_info:
        if disk['left'] and disk['right']:
            plt.plot([disk['left'][0], disk['right'][0]], [disk['left'][1], disk['right'][1]], 'r-', linewidth=2)
        plt.scatter(disk['center'][0], disk['center'][1], c='yellow', s=20)
    plt.title(f'{view} {phase} - Measurement Disks')
    
    # Volume contribution per disk
    plt.subplot(133)
    if disk_info and 'diameter_cm' in disk_info[0]:
        diameters = [d['diameter_cm'] for d in disk_info]
        plt.plot(diameters, 'bo-')
        plt.title(f'{view} {phase} - Disk Diameters')
        plt.xlabel('Disk Number')
        plt.ylabel('Diameter (cm)')
        plt.grid(True)
    
    plt.tight_layout()
    plt.show()

def process_patient(patient_dir, visualize=False):
    """Process one patient with improved volume calculation"""
    patient_id = os.path.basename(patient_dir)
    print(f"\nProcessing Patient: {patient_id}")
    
    volumes = {'2CH': {'ED': None, 'ES': None}, '4CH': {'ED': None, 'ES': None}}
    all_landmarks = {}
    all_disk_info = {}
    
    for file in os.listdir(patient_dir):
        if not (file.endswith('_gt.niigz') or file.endswith('_gt.nii.gz')):
            continue
            
        try:
            # Determine view and phase
            view = '4CH' if '4ch' in file.lower() else '2CH' if '2ch' in file.lower() else None
            phase = 'ED' if 'ed' in file.lower() else 'ES' if 'es' in file.lower() else None
            
            if not view or not phase:
                continue
                
            print(f"\nProcessing {view} {phase}...")
            
            # Handle compressed files
            filepath = os.path.join(patient_dir, file)
            temp_file = None
            
            if file.endswith('.niigz'):
                temp_file = handle_compressed_file(filepath)
                gt_img = nib.load(temp_file)
            else:
                gt_img = nib.load(filepath)
                
            gt_data = gt_img.get_fdata()
            pixel_spacing = gt_img.header.get_zooms()[:2]
            
            # Create mask and rotate
            mask = np.zeros_like(gt_data, dtype=np.uint8)
            mask[gt_data == 1] = 255
            rotated_mask = rotate_image_clockwise(mask)
            
            # Detect landmarks
            apex, septal, lateral, mitral_mid = detect_landmarks(rotated_mask)
            
            # Calculate disk diameters and dead space
            L_cm, diameters_cm, disk_info, dead_area_cm2 = compute_disk_diameters(
                rotated_mask, apex, mitral_mid, pixel_spacing
            )
            
            # Store results
            volumes[view][phase] = {
                'volume': None,  # Will be calculated after both views are processed
                'L_cm': L_cm,
                'diameters_cm': diameters_cm,
                'dead_area_cm2': dead_area_cm2,
                'status': 'success'
            }
            
            all_landmarks[(view, phase)] = (apex, septal, lateral, mitral_mid)
            all_disk_info[(view, phase)] = disk_info
            
            # Visualize if requested
            if visualize:
                visualize_processing(rotated_mask, (apex, septal, lateral, mitral_mid), 
                                  disk_info, view, phase, pixel_spacing)
            
        except Exception as e:
            print(f"Error processing {file}: {str(e)}")
            if view and phase:
                volumes[view][phase] = {
                    'volume': None,
                    'status': f'failed: {str(e)}'
                }
        finally:
            if temp_file and os.path.exists(temp_file):
                os.unlink(temp_file)
    
    # Calculate volumes and EF if we have all required data
    missing = []
    for view in ['2CH', '4CH']:
        for phase in ['ED', 'ES']:
            if not volumes[view][phase] or volumes[view][phase]['status'] != 'success':
                missing.append(f"{view} {phase}")
    
    if missing:
        print(f"\nIncomplete data for EF calculation. Missing: {', '.join(missing)}")
    else:
        # Calculate EDV and ESV using biplane Simpson's method
        try:
            # ED volumes
            ed_vol = compute_biplane_volume(
                volumes['4CH']['ED']['L_cm'],
                volumes['2CH']['ED']['L_cm'],
                volumes['4CH']['ED']['diameters_cm'],
                volumes['2CH']['ED']['diameters_cm'],
                volumes['4CH']['ED']['dead_area_cm2'],
                volumes['2CH']['ED']['dead_area_cm2']
            )
            
            # ES volumes
            es_vol = compute_biplane_volume(
                volumes['4CH']['ES']['L_cm'],
                volumes['2CH']['ES']['L_cm'],
                volumes['4CH']['ES']['diameters_cm'],
                volumes['2CH']['ES']['diameters_cm'],
                volumes['4CH']['ES']['dead_area_cm2'],
                volumes['2CH']['ES']['dead_area_cm2']
            )
            
            # Calculate EF
            ef = ((ed_vol - es_vol) / ed_vol) * 100 if ed_vol > 0 else 0
            
            # Update volumes dictionary
            volumes['EDV'] = ed_vol
            volumes['ESV'] = es_vol
            volumes['EF'] = ef
            
            print(f"\nEF Calculation Successful:")
            print(f"EDV: {ed_vol:.2f} ml")
            print(f"ESV: {es_vol:.2f} ml")
            print(f"EF: {ef:.2f}%")
            
            # Final visualization
            if visualize:
                plt.figure(figsize=(10, 5))
                bars = plt.bar(['EDV', 'ESV'], [ed_vol, es_vol], color=['blue', 'red'])
                plt.ylabel('Volume (ml)')
                plt.title(f'Ejection Fraction Calculation\nEF = {ef:.1f}%')
                for bar in bars:
                    height = bar.get_height()
                    plt.text(bar.get_x() + bar.get_width()/2., height,
                            f'{height:.1f} ml', ha='center', va='bottom')
                plt.show()
                
        except Exception as e:
            print(f"Error calculating volumes: {str(e)}")
    
    return volumes, all_landmarks, all_disk_info

# Process one patient
base_dir = r"E:\echo heart\Resources"
patient_dirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

if patient_dirs:
    patient_volumes, landmarks, disk_info = process_patient(
        os.path.join(base_dir, patient_dirs[3]), 
        visualize=True
    )
else:
    print("No patient directories found")
```

```python
import os
import numpy as np
import nibabel as nib
import cv2
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
import math
import gzip
import shutil
import tempfile
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

def handle_compressed_file(filepath):
    """Handle both .nii.gz and .niigz files"""
    if filepath.endswith('.niigz'):
        with tempfile.NamedTemporaryFile(suffix='.nii.gz', delete=False) as tmp_file:
            with gzip.open(filepath, 'rb') as f_in:
                shutil.copyfileobj(f_in, tmp_file)
            return tmp_file.name
    return filepath

def rotate_image_clockwise(image, angle=90):
    """Rotate image clockwise with better boundary handling"""
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, -angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_NEAREST)
    return rotated

def detect_landmarks(seg_binary, delta_y_fraction=0.1, y_tolerance=10):
    """Improved landmark detection with curvature-based refinement."""
    # Convert to binary and find contours
    seg_binary = (seg_binary > 0).astype(np.uint8) * 255
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        raise ValueError("No LV contour found")
    
    # Use the largest contour
    contour = max(contours, key=cv2.contourArea).squeeze()
    x = contour[:, 0].astype(float)
    y = contour[:, 1].astype(float)
    
    # Smooth contour
    x_smooth = gaussian_filter1d(x, sigma=5, mode='wrap')
    y_smooth = gaussian_filter1d(y, sigma=5, mode='wrap')
    
    # Compute curvature
    dx = np.gradient(x_smooth)
    dy = np.gradient(y_smooth)
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(dx * ddy - dy * ddx) / (dx**2 + dy**2 + 1e-10)**1.5
    curvature = (curvature - np.min(curvature)) / (np.max(curvature) - np.min(curvature) + 1e-10)
    
    # Find apex: min y
    apex_idx = np.argmin(y)
    apex_point = (int(x[apex_idx]), int(y[apex_idx]))
    
    # Find all peaks in curvature
    peaks, _ = find_peaks(curvature, prominence=0.0005)
    
    # Define basal region: y > max_y * (1 - delta_y_fraction)
    max_y = max(y)
    min_y = min(y)
    basal_threshold = max_y - delta_y_fraction * (max_y - min_y)
    basal_peaks = [i for i in peaks if y[i] >= basal_threshold]
    
    if len(basal_peaks) >= 2:
        # Sort peaks by curvature (highest to lowest)
        sorted_peaks = sorted(basal_peaks, key=lambda i: curvature[i], reverse=True)
        # Assign first highest peak as septal, second as lateral
        septal_idx = sorted_peaks[0]
        lateral_idx = sorted_peaks[1]
        
        # Verify they are on one line (similar y-coordinates)
        if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
            # Find a lateral peak with similar y-coordinate to septal
            septal_y = y[septal_idx]
            lateral_candidates = [i for i in basal_peaks if abs(y[i] - septal_y) <= y_tolerance]
            if len(lateral_candidates) >= 2:
                # Select the highest curvature peak among candidates, excluding septal
                lateral_candidates = [i for i in lateral_candidates if i != septal_idx]
                if lateral_candidates:
                    lateral_idx = max(lateral_candidates, key=lambda i: curvature[i])
                else:
                    raise ValueError("No suitable lateral peak found with similar y-coordinate")
            else:
                raise ValueError("Not enough peaks with similar y-coordinates")
    else:
        # Fallback: points with y > mean(y), min and max x
        mean_y = np.mean(y)
        lower_indices = [i for i in range(len(y)) if y[i] > mean_y]
        if len(lower_indices) >= 2:
            septal_idx = min(lower_indices, key=lambda i: x[i])
            lateral_idx = max(lower_indices, key=lambda i: x[i])
            # Verify y-alignment in fallback
            if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
                raise ValueError("Fallback septal and lateral points not on one line")
        else:
            raise ValueError("Not enough points for fallback")
    
    septal_point = (int(x[septal_idx]), int(y[septal_idx]))
    lateral_point = (int(x[lateral_idx]), int(y[lateral_idx]))
    
    # Mitral mid: midpoint between septal and lateral
    mitral_mid = ((septal_point[0] + lateral_point[0]) / 2, (septal_point[1] + lateral_point[1]) / 2)
    
    return apex_point, septal_point, lateral_point, mitral_mid

def compute_disk_diameters(seg_binary, apex, mitral_mid, pixel_spacing=(0.1, 0.1), nr_disks=20):
    """Compute disk diameters, long axis length, and dead space for a single view."""
    # Precompute filled LV mask
    contours, _ = cv2.findContours(seg_binary.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        return 0.0, [], [], 0.0
    
    max_contour = max(contours, key=cv2.contourArea)
    filled_mask = np.zeros_like(seg_binary)
    cv2.drawContours(filled_mask, [max_contour], -1, 1, thickness=cv2.FILLED)
    
    # Compute boundary
    kernel = np.ones((3,3), np.uint8)
    dilation = cv2.dilate(filled_mask, kernel)
    erosion = cv2.erode(filled_mask, kernel)
    boundary = (dilation - erosion > 0).astype(np.uint8)
    
    # Long axis vector
    dx = apex[0] - mitral_mid[0]
    dy = apex[1] - mitral_mid[1]
    L_px = np.sqrt(dx**2 + dy**2)
    L_cm = L_px * pixel_spacing[0]  # Assume isotropic spacing
    
    if L_cm == 0:
        return L_cm, [], [], 0.0
    
    # Perpendicular vector
    perp_vector = np.array([-dy, dx]) / L_px if L_px > 0 else np.array([1, 0])
    
    diameters_cm = []
    disk_info = []
    
    for i in range(nr_disks):
        t = (i + 0.5) / nr_disks
        center_x = mitral_mid[0] + t * dx
        center_y = mitral_mid[1] + t * dy
        
        left_point = None
        right_point = None
        
        # Search in both directions
        for direction in [-1, 1]:
            for dist in range(1, 150):
                px = int(round(center_x + direction * dist * perp_vector[0]))
                py = int(round(center_y + direction * dist * perp_vector[1]))
                
                if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                    break
                
                if boundary[py, px]:
                    if direction == -1:
                        left_point = (px, py)
                    else:
                        right_point = (px, py)
                    break
        
        if left_point and right_point:
            dx_px = right_point[0] - left_point[0]
            dy_px = right_point[1] - left_point[1]
            diameter_cm = np.sqrt((dx_px * pixel_spacing[0])**2 + (dy_px * pixel_spacing[1])**2)
            diameters_cm.append(diameter_cm)
            disk_info.append({
                'center': (center_x, center_y),
                'left': left_point,
                'right': right_point,
                'diameter_cm': diameter_cm
            })
        else:
            diameters_cm.append(0.0)
            disk_info.append({
                'center': (center_x, center_y),
                'left': None,
                'right': None,
                'diameter_cm': 0.0
            })
    
    # Calculate dead space area
    dead_area_cm2 = 0.0
    
    if len([d for d in disk_info if d['left'] and d['right']]) >= 2:
        # Create polygon from disk endpoints
        polygon_points = []
        
        # Add left points from base to apex
        for disk in disk_info:
            if disk['left']:
                polygon_points.append(disk['left'])
        
        # Add right points from apex to base
        for disk in reversed(disk_info):
            if disk['right']:
                polygon_points.append(disk['right'])
        
        # Create mask for disk-covered area
        disk_area_mask = np.zeros_like(filled_mask, dtype=np.uint8)
        if len(polygon_points) > 2:
            pts = np.array([polygon_points], dtype=np.int32)
            cv2.fillPoly(disk_area_mask, [pts], 1)
            
            # Calculate dead space as LV area not covered by disks
            dead_space_mask = filled_mask - disk_area_mask
            dead_space_mask = np.clip(dead_space_mask, 0, 1)  # Ensure binary
            
            # Convert to cm
            dead_area_px = np.sum(dead_space_mask)
            dead_area_cm2 = dead_area_px * pixel_spacing[0] * pixel_spacing[1]
    
    return L_cm, diameters_cm, disk_info, dead_area_cm2

def compute_biplane_volume(L4_cm, L2_cm, diameters4_cm, diameters2_cm, dead_area4_cm2, dead_area2_cm2, nr_disks=20):
    """Compute biplane Simpson's volume with dead space correction."""
    # Use average long axis length
    L_avg_cm = (L4_cm + L2_cm) / 2
    h_cm = L_avg_cm / nr_disks
    
    # Simpson's volume from disks
    total_volume = 0.0
    disk_volumes = []
    
    for i in range(nr_disks):
        a_i = diameters4_cm[i] if i < len(diameters4_cm) else 0
        b_i = diameters2_cm[i] if i < len(diameters2_cm) else 0
        disk_volume = (np.pi / 4) * a_i * b_i * h_cm
        total_volume += disk_volume
        disk_volumes.append(disk_volume)
    
    # Calculate dead volume
    avg_dead_area_cm2 = (dead_area4_cm2 + dead_area2_cm2) / 2
    dead_volume = avg_dead_area_cm2 * h_cm
    
    # Add dead volume to total
    total_volume_with_dead = total_volume + dead_volume
    
    return total_volume_with_dead

def visualize_processing(seg_data, landmarks, disk_info, view, phase, pixel_spacing):
    """Visualize the processing steps for one view/phase"""
    apex, septal, lateral, mitral_mid = landmarks
    
    # Create figure
    plt.figure(figsize=(15, 5))
    
    # Original image with landmarks
    plt.subplot(131)
    plt.imshow(seg_data, cmap='gray')
    plt.scatter(apex[0], apex[1], c='red', s=80, marker='x', label='Apex')
    plt.scatter(septal[0], septal[1], c='blue', s=80, marker='x', label='Septal')
    plt.scatter(lateral[0], lateral[1], c='green', s=80, marker='x', label='Lateral')
    plt.scatter(mitral_mid[0], mitral_mid[1], c='magenta', s=100, marker='+', label='Mitral Mid')
    plt.plot([apex[0], mitral_mid[0]], [apex[1], mitral_mid[1]], 'y--', label='Long Axis')
    plt.plot([septal[0], lateral[0]], [septal[1], lateral[1]], 'c--', label='Mitral Line')
    plt.title(f'{view} {phase} - Landmarks')
    plt.legend()
    
    # Measurement disks
    plt.subplot(132)
    plt.imshow(seg_data, cmap='gray')
    for disk in disk_info:
        if disk['left'] and disk['right']:
            plt.plot([disk['left'][0], disk['right'][0]], [disk['left'][1], disk['right'][1]], 'r-', linewidth=2)
        plt.scatter(disk['center'][0], disk['center'][1], c='yellow', s=20)
    plt.title(f'{view} {phase} - Measurement Disks')
    
    # Volume contribution per disk
    plt.subplot(133)
    if disk_info and 'diameter_cm' in disk_info[0]:
        diameters = [d['diameter_cm'] for d in disk_info]
        plt.plot(diameters, 'bo-')
        plt.title(f'{view} {phase} - Disk Diameters')
        plt.xlabel('Disk Number')
        plt.ylabel('Diameter (cm)')
        plt.grid(True)
    
    plt.tight_layout()
    plt.show()

def process_patient(patient_dir, visualize=False):
    """Process one patient with improved volume calculation"""
    patient_id = os.path.basename(patient_dir)
    print(f"\nProcessing Patient: {patient_id}")
    
    volumes = {'2CH': {'ED': None, 'ES': None}, '4CH': {'ED': None, 'ES': None}}
    all_landmarks = {}
    all_disk_info = {}
    
    for file in os.listdir(patient_dir):
        if not (file.endswith('_gt.niigz') or file.endswith('_gt.nii.gz')):
            continue
            
        try:
            # Determine view and phase
            view = '4CH' if '4ch' in file.lower() else '2CH' if '2ch' in file.lower() else None
            phase = 'ED' if 'ed' in file.lower() else 'ES' if 'es' in file.lower() else None
            
            if not view or not phase:
                continue
                
            print(f"\nProcessing {view} {phase}...")
            
            # Handle compressed files
            filepath = os.path.join(patient_dir, file)
            temp_file = None
            
            if file.endswith('.niigz'):
                temp_file = handle_compressed_file(filepath)
                gt_img = nib.load(temp_file)
            else:
                gt_img = nib.load(filepath)
                
            gt_data = gt_img.get_fdata()
            pixel_spacing = gt_img.header.get_zooms()[:2]
            
            # Create mask and rotate
            mask = np.zeros_like(gt_data, dtype=np.uint8)
            mask[gt_data == 1] = 255
            rotated_mask = rotate_image_clockwise(mask)
            
            # Detect landmarks
            apex, septal, lateral, mitral_mid = detect_landmarks(rotated_mask)
            
            # Calculate disk diameters and dead space
            L_cm, diameters_cm, disk_info, dead_area_cm2 = compute_disk_diameters(
                rotated_mask, apex, mitral_mid, pixel_spacing
            )
            
            # Store results
            volumes[view][phase] = {
                'volume': None,  # Will be calculated after both views are processed
                'L_cm': L_cm,
                'diameters_cm': diameters_cm,
                'dead_area_cm2': dead_area_cm2,
                'status': 'success'
            }
            
            all_landmarks[(view, phase)] = (apex, septal, lateral, mitral_mid)
            all_disk_info[(view, phase)] = disk_info
            
            # Visualize if requested
            if visualize:
                visualize_processing(rotated_mask, (apex, septal, lateral, mitral_mid), 
                                  disk_info, view, phase, pixel_spacing)
            
        except Exception as e:
            print(f"Error processing {file}: {str(e)}")
            if view and phase:
                volumes[view][phase] = {
                    'volume': None,
                    'status': f'failed: {str(e)}'
                }
        finally:
            if temp_file and os.path.exists(temp_file):
                os.unlink(temp_file)
    
    # Calculate volumes and EF if we have all required data
    missing = []
    for view in ['2CH', '4CH']:
        for phase in ['ED', 'ES']:
            if not volumes[view][phase] or volumes[view][phase]['status'] != 'success':
                missing.append(f"{view} {phase}")
    
    if missing:
        print(f"\nIncomplete data for EF calculation. Missing: {', '.join(missing)}")
    else:
        # Calculate EDV and ESV using biplane Simpson's method
        try:
            # ED volumes
            ed_vol = compute_biplane_volume(
                volumes['4CH']['ED']['L_cm'],
                volumes['2CH']['ED']['L_cm'],
                volumes['4CH']['ED']['diameters_cm'],
                volumes['2CH']['ED']['diameters_cm'],
                volumes['4CH']['ED']['dead_area_cm2'],
                volumes['2CH']['ED']['dead_area_cm2']
            )
            
            # ES volumes
            es_vol = compute_biplane_volume(
                volumes['4CH']['ES']['L_cm'],
                volumes['2CH']['ES']['L_cm'],
                volumes['4CH']['ES']['diameters_cm'],
                volumes['2CH']['ES']['diameters_cm'],
                volumes['4CH']['ES']['dead_area_cm2'],
                volumes['2CH']['ES']['dead_area_cm2']
            )
            
            # Calculate EF
            ef = ((ed_vol - es_vol) / ed_vol) * 100 if ed_vol > 0 else 0
            
            # Update volumes dictionary
            volumes['EDV'] = ed_vol
            volumes['ESV'] = es_vol
            volumes['EF'] = ef
            
            print(f"\nEF Calculation Successful:")
            print(f"EDV: {ed_vol:.2f} ml")
            print(f"ESV: {es_vol:.2f} ml")
            print(f"EF: {ef:.2f}%")
            
            # Final visualization
            if visualize:
                plt.figure(figsize=(10, 5))
                bars = plt.bar(['EDV', 'ESV'], [ed_vol, es_vol], color=['blue', 'red'])
                plt.ylabel('Volume (ml)')
                plt.title(f'Ejection Fraction Calculation\nEF = {ef:.1f}%')
                for bar in bars:
                    height = bar.get_height()
                    plt.text(bar.get_x() + bar.get_width()/2., height,
                            f'{height:.1f} ml', ha='center', va='bottom')
                plt.show()
                
        except Exception as e:
            print(f"Error calculating volumes: {str(e)}")
    
    return volumes, all_landmarks, all_disk_info

# Process one patient
base_dir = r"E:\echo heart\Resources"
patient_dirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

if patient_dirs:
    patient_volumes, landmarks, disk_info = process_patient(
        os.path.join(base_dir, patient_dirs[6]), 
        visualize=True
    )
else:
    print("No patient directories found")
```

```python
import os
import numpy as np
import nibabel as nib
import cv2
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
import math
import gzip
import shutil
import tempfile
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

def handle_compressed_file(filepath):
    """Handle both .nii.gz and .niigz files"""
    if filepath.endswith('.niigz'):
        with tempfile.NamedTemporaryFile(suffix='.nii.gz', delete=False) as tmp_file:
            with gzip.open(filepath, 'rb') as f_in:
                shutil.copyfileobj(f_in, tmp_file)
            return tmp_file.name
    return filepath

def rotate_image_clockwise(image, angle=90):
    """Rotate image clockwise with better boundary handling"""
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, -angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_NEAREST)
    return rotated

def detect_landmarks(seg_binary, delta_y_fraction=0.1, y_tolerance=10):
    """Improved landmark detection with curvature-based refinement."""
    # Convert to binary and find contours
    seg_binary = (seg_binary > 0).astype(np.uint8) * 255
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        raise ValueError("No LV contour found")
    
    # Use the largest contour
    contour = max(contours, key=cv2.contourArea).squeeze()
    x = contour[:, 0].astype(float)
    y = contour[:, 1].astype(float)
    
    # Smooth contour
    x_smooth = gaussian_filter1d(x, sigma=5, mode='wrap')
    y_smooth = gaussian_filter1d(y, sigma=5, mode='wrap')
    
    # Compute curvature
    dx = np.gradient(x_smooth)
    dy = np.gradient(y_smooth)
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(dx * ddy - dy * ddx) / (dx**2 + dy**2 + 1e-10)**1.5
    curvature = (curvature - np.min(curvature)) / (np.max(curvature) - np.min(curvature) + 1e-10)
    
    # Find apex: min y
    apex_idx = np.argmin(y)
    apex_point = (int(x[apex_idx]), int(y[apex_idx]))
    
    # Find all peaks in curvature
    peaks, _ = find_peaks(curvature, prominence=0.0005)
    
    # Define basal region: y > max_y * (1 - delta_y_fraction)
    max_y = max(y)
    min_y = min(y)
    basal_threshold = max_y - delta_y_fraction * (max_y - min_y)
    basal_peaks = [i for i in peaks if y[i] >= basal_threshold]
    
    if len(basal_peaks) >= 2:
        # Sort peaks by curvature (highest to lowest)
        sorted_peaks = sorted(basal_peaks, key=lambda i: curvature[i], reverse=True)
        # Assign first highest peak as septal, second as lateral
        septal_idx = sorted_peaks[0]
        lateral_idx = sorted_peaks[1]
        
        # Verify they are on one line (similar y-coordinates)
        if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
            # Find a lateral peak with similar y-coordinate to septal
            septal_y = y[septal_idx]
            lateral_candidates = [i for i in basal_peaks if abs(y[i] - septal_y) <= y_tolerance]
            if len(lateral_candidates) >= 2:
                # Select the highest curvature peak among candidates, excluding septal
                lateral_candidates = [i for i in lateral_candidates if i != septal_idx]
                if lateral_candidates:
                    lateral_idx = max(lateral_candidates, key=lambda i: curvature[i])
                else:
                    raise ValueError("No suitable lateral peak found with similar y-coordinate")
            else:
                raise ValueError("Not enough peaks with similar y-coordinates")
    else:
        # Fallback: points with y > mean(y), min and max x
        mean_y = np.mean(y)
        lower_indices = [i for i in range(len(y)) if y[i] > mean_y]
        if len(lower_indices) >= 2:
            septal_idx = min(lower_indices, key=lambda i: x[i])
            lateral_idx = max(lower_indices, key=lambda i: x[i])
            # Verify y-alignment in fallback
            if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
                raise ValueError("Fallback septal and lateral points not on one line")
        else:
            raise ValueError("Not enough points for fallback")
    
    septal_point = (int(x[septal_idx]), int(y[septal_idx]))
    lateral_point = (int(x[lateral_idx]), int(y[lateral_idx]))
    
    # Mitral mid: midpoint between septal and lateral
    mitral_mid = ((septal_point[0] + lateral_point[0]) / 2, (septal_point[1] + lateral_point[1]) / 2)
    
    return apex_point, septal_point, lateral_point, mitral_mid

def compute_disk_diameters(seg_binary, apex, mitral_mid, pixel_spacing=(0.1, 0.1), nr_disks=20):
    """Compute disk diameters, long axis length, and dead space for a single view."""
    # Precompute filled LV mask
    contours, _ = cv2.findContours(seg_binary.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        return 0.0, [], [], 0.0
    
    max_contour = max(contours, key=cv2.contourArea)
    filled_mask = np.zeros_like(seg_binary)
    cv2.drawContours(filled_mask, [max_contour], -1, 1, thickness=cv2.FILLED)
    
    # Compute boundary
    kernel = np.ones((3,3), np.uint8)
    dilation = cv2.dilate(filled_mask, kernel)
    erosion = cv2.erode(filled_mask, kernel)
    boundary = (dilation - erosion > 0).astype(np.uint8)
    
    # Long axis vector
    dx = apex[0] - mitral_mid[0]
    dy = apex[1] - mitral_mid[1]
    L_px = np.sqrt(dx**2 + dy**2)
    L_cm = L_px * pixel_spacing[0]  # Assume isotropic spacing
    
    if L_cm == 0:
        return L_cm, [], [], 0.0
    
    # Perpendicular vector
    perp_vector = np.array([-dy, dx]) / L_px if L_px > 0 else np.array([1, 0])
    
    diameters_cm = []
    disk_info = []
    
    for i in range(nr_disks):
        t = (i + 0.5) / nr_disks
        center_x = mitral_mid[0] + t * dx
        center_y = mitral_mid[1] + t * dy
        
        left_point = None
        right_point = None
        
        # Search in both directions
        for direction in [-1, 1]:
            for dist in range(1, 150):
                px = int(round(center_x + direction * dist * perp_vector[0]))
                py = int(round(center_y + direction * dist * perp_vector[1]))
                
                if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                    break
                
                if boundary[py, px]:
                    if direction == -1:
                        left_point = (px, py)
                    else:
                        right_point = (px, py)
                    break
        
        if left_point and right_point:
            dx_px = right_point[0] - left_point[0]
            dy_px = right_point[1] - left_point[1]
            diameter_cm = np.sqrt((dx_px * pixel_spacing[0])**2 + (dy_px * pixel_spacing[1])**2)
            diameters_cm.append(diameter_cm)
            disk_info.append({
                'center': (center_x, center_y),
                'left': left_point,
                'right': right_point,
                'diameter_cm': diameter_cm
            })
        else:
            diameters_cm.append(0.0)
            disk_info.append({
                'center': (center_x, center_y),
                'left': None,
                'right': None,
                'diameter_cm': 0.0
            })
    
    # Calculate dead space area
    dead_area_cm2 = 0.0
    
    if len([d for d in disk_info if d['left'] and d['right']]) >= 2:
        # Create polygon from disk endpoints
        polygon_points = []
        
        # Add left points from base to apex
        for disk in disk_info:
            if disk['left']:
                polygon_points.append(disk['left'])
        
        # Add right points from apex to base
        for disk in reversed(disk_info):
            if disk['right']:
                polygon_points.append(disk['right'])
        
        # Create mask for disk-covered area
        disk_area_mask = np.zeros_like(filled_mask, dtype=np.uint8)
        if len(polygon_points) > 2:
            pts = np.array([polygon_points], dtype=np.int32)
            cv2.fillPoly(disk_area_mask, [pts], 1)
            
            # Calculate dead space as LV area not covered by disks
            dead_space_mask = filled_mask - disk_area_mask
            dead_space_mask = np.clip(dead_space_mask, 0, 1)  # Ensure binary
            
            # Convert to cm
            dead_area_px = np.sum(dead_space_mask)
            dead_area_cm2 = dead_area_px * pixel_spacing[0] * pixel_spacing[1]
    
    return L_cm, diameters_cm, disk_info, dead_area_cm2

def compute_biplane_volume(L4_cm, L2_cm, diameters4_cm, diameters2_cm, dead_area4_cm2, dead_area2_cm2, nr_disks=20):
    """Compute biplane Simpson's volume with dead space correction."""
    # Use average long axis length
    L_avg_cm = (L4_cm + L2_cm) / 2
    h_cm = L_avg_cm / nr_disks
    
    # Simpson's volume from disks
    total_volume = 0.0
    disk_volumes = []
    
    for i in range(nr_disks):
        a_i = diameters4_cm[i] if i < len(diameters4_cm) else 0
        b_i = diameters2_cm[i] if i < len(diameters2_cm) else 0
        disk_volume = (np.pi / 4) * a_i * b_i * h_cm
        total_volume += disk_volume
        disk_volumes.append(disk_volume)
    
    # Calculate dead volume
    avg_dead_area_cm2 = (dead_area4_cm2 + dead_area2_cm2) / 2
    dead_volume = avg_dead_area_cm2 * h_cm
    
    # Add dead volume to total
    total_volume_with_dead = total_volume + dead_volume
    
    return total_volume_with_dead

def visualize_processing(seg_data, landmarks, disk_info, view, phase, pixel_spacing):
    """Visualize the processing steps for one view/phase"""
    apex, septal, lateral, mitral_mid = landmarks
    
    # Create figure
    plt.figure(figsize=(15, 5))
    
    # Original image with landmarks
    plt.subplot(131)
    plt.imshow(seg_data, cmap='gray')
    plt.scatter(apex[0], apex[1], c='red', s=80, marker='x', label='Apex')
    plt.scatter(septal[0], septal[1], c='blue', s=80, marker='x', label='Septal')
    plt.scatter(lateral[0], lateral[1], c='green', s=80, marker='x', label='Lateral')
    plt.scatter(mitral_mid[0], mitral_mid[1], c='magenta', s=100, marker='+', label='Mitral Mid')
    plt.plot([apex[0], mitral_mid[0]], [apex[1], mitral_mid[1]], 'y--', label='Long Axis')
    plt.plot([septal[0], lateral[0]], [septal[1], lateral[1]], 'c--', label='Mitral Line')
    plt.title(f'{view} {phase} - Landmarks')
    plt.legend()
    
    # Measurement disks
    plt.subplot(132)
    plt.imshow(seg_data, cmap='gray')
    for disk in disk_info:
        if disk['left'] and disk['right']:
            plt.plot([disk['left'][0], disk['right'][0]], [disk['left'][1], disk['right'][1]], 'r-', linewidth=2)
        plt.scatter(disk['center'][0], disk['center'][1], c='yellow', s=20)
    plt.title(f'{view} {phase} - Measurement Disks')
    
    # Volume contribution per disk
    plt.subplot(133)
    if disk_info and 'diameter_cm' in disk_info[0]:
        diameters = [d['diameter_cm'] for d in disk_info]
        plt.plot(diameters, 'bo-')
        plt.title(f'{view} {phase} - Disk Diameters')
        plt.xlabel('Disk Number')
        plt.ylabel('Diameter (cm)')
        plt.grid(True)
    
    plt.tight_layout()
    plt.show()

def process_patient(patient_dir, visualize=False):
    """Process one patient with improved volume calculation"""
    patient_id = os.path.basename(patient_dir)
    print(f"\nProcessing Patient: {patient_id}")
    
    volumes = {'2CH': {'ED': None, 'ES': None}, '4CH': {'ED': None, 'ES': None}}
    all_landmarks = {}
    all_disk_info = {}
    
    for file in os.listdir(patient_dir):
        if not (file.endswith('_gt.niigz') or file.endswith('_gt.nii.gz')):
            continue
            
        try:
            # Determine view and phase
            view = '4CH' if '4ch' in file.lower() else '2CH' if '2ch' in file.lower() else None
            phase = 'ED' if 'ed' in file.lower() else 'ES' if 'es' in file.lower() else None
            
            if not view or not phase:
                continue
                
            print(f"\nProcessing {view} {phase}...")
            
            # Handle compressed files
            filepath = os.path.join(patient_dir, file)
            temp_file = None
            
            if file.endswith('.niigz'):
                temp_file = handle_compressed_file(filepath)
                gt_img = nib.load(temp_file)
            else:
                gt_img = nib.load(filepath)
                
            gt_data = gt_img.get_fdata()
            pixel_spacing = gt_img.header.get_zooms()[:2]
            
            # Create mask and rotate
            mask = np.zeros_like(gt_data, dtype=np.uint8)
            mask[gt_data == 1] = 255
            rotated_mask = rotate_image_clockwise(mask)
            
            # Detect landmarks
            apex, septal, lateral, mitral_mid = detect_landmarks(rotated_mask)
            
            # Calculate disk diameters and dead space
            L_cm, diameters_cm, disk_info, dead_area_cm2 = compute_disk_diameters(
                rotated_mask, apex, mitral_mid, pixel_spacing
            )
            
            # Store results
            volumes[view][phase] = {
                'volume': None,  # Will be calculated after both views are processed
                'L_cm': L_cm,
                'diameters_cm': diameters_cm,
                'dead_area_cm2': dead_area_cm2,
                'status': 'success'
            }
            
            all_landmarks[(view, phase)] = (apex, septal, lateral, mitral_mid)
            all_disk_info[(view, phase)] = disk_info
            
            # Visualize if requested
            if visualize:
                visualize_processing(rotated_mask, (apex, septal, lateral, mitral_mid), 
                                  disk_info, view, phase, pixel_spacing)
            
        except Exception as e:
            print(f"Error processing {file}: {str(e)}")
            if view and phase:
                volumes[view][phase] = {
                    'volume': None,
                    'status': f'failed: {str(e)}'
                }
        finally:
            if temp_file and os.path.exists(temp_file):
                os.unlink(temp_file)
    
    # Calculate volumes and EF if we have all required data
    missing = []
    for view in ['2CH', '4CH']:
        for phase in ['ED', 'ES']:
            if not volumes[view][phase] or volumes[view][phase]['status'] != 'success':
                missing.append(f"{view} {phase}")
    
    if missing:
        print(f"\nIncomplete data for EF calculation. Missing: {', '.join(missing)}")
    else:
        # Calculate EDV and ESV using biplane Simpson's method
        try:
            # ED volumes
            ed_vol = compute_biplane_volume(
                volumes['4CH']['ED']['L_cm'],
                volumes['2CH']['ED']['L_cm'],
                volumes['4CH']['ED']['diameters_cm'],
                volumes['2CH']['ED']['diameters_cm'],
                volumes['4CH']['ED']['dead_area_cm2'],
                volumes['2CH']['ED']['dead_area_cm2']
            )
            
            # ES volumes
            es_vol = compute_biplane_volume(
                volumes['4CH']['ES']['L_cm'],
                volumes['2CH']['ES']['L_cm'],
                volumes['4CH']['ES']['diameters_cm'],
                volumes['2CH']['ES']['diameters_cm'],
                volumes['4CH']['ES']['dead_area_cm2'],
                volumes['2CH']['ES']['dead_area_cm2']
            )
            
            # Calculate EF
            ef = ((ed_vol - es_vol) / ed_vol) * 100 if ed_vol > 0 else 0
            
            # Update volumes dictionary
            volumes['EDV'] = ed_vol
            volumes['ESV'] = es_vol
            volumes['EF'] = ef
            
            print(f"\nEF Calculation Successful:")
            print(f"EDV: {ed_vol:.2f} ml")
            print(f"ESV: {es_vol:.2f} ml")
            print(f"EF: {ef:.2f}%")
            
            # Final visualization
            if visualize:
                plt.figure(figsize=(10, 5))
                bars = plt.bar(['EDV', 'ESV'], [ed_vol, es_vol], color=['blue', 'red'])
                plt.ylabel('Volume (ml)')
                plt.title(f'Ejection Fraction Calculation\nEF = {ef:.1f}%')
                for bar in bars:
                    height = bar.get_height()
                    plt.text(bar.get_x() + bar.get_width()/2., height,
                            f'{height:.1f} ml', ha='center', va='bottom')
                plt.show()
                
        except Exception as e:
            print(f"Error calculating volumes: {str(e)}")
    
    return volumes, all_landmarks, all_disk_info

# Process one patient
base_dir = r"E:\echo heart\Resources"
patient_dirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

if patient_dirs:
    patient_volumes, landmarks, disk_info = process_patient(
        os.path.join(base_dir, patient_dirs[9]), 
        visualize=True
    )
else:
    print("No patient directories found")
```

```python
import os
import numpy as np
import nibabel as nib
import cv2
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
import math
import gzip
import shutil
import tempfile
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

def handle_compressed_file(filepath):
    """Handle both .nii.gz and .niigz files"""
    if filepath.endswith('.niigz'):
        with tempfile.NamedTemporaryFile(suffix='.nii.gz', delete=False) as tmp_file:
            with gzip.open(filepath, 'rb') as f_in:
                shutil.copyfileobj(f_in, tmp_file)
            return tmp_file.name
    return filepath

def rotate_image_clockwise(image, angle=90):
    """Rotate image clockwise with better boundary handling"""
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, -angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_NEAREST)
    return rotated

def detect_landmarks(seg_binary, delta_y_fraction=0.1, y_tolerance=10):
    """Improved landmark detection with curvature-based refinement."""
    # Convert to binary and find contours
    seg_binary = (seg_binary > 0).astype(np.uint8) * 255
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        raise ValueError("No LV contour found")
    
    # Use the largest contour
    contour = max(contours, key=cv2.contourArea).squeeze()
    x = contour[:, 0].astype(float)
    y = contour[:, 1].astype(float)
    
    # Smooth contour
    x_smooth = gaussian_filter1d(x, sigma=5, mode='wrap')
    y_smooth = gaussian_filter1d(y, sigma=5, mode='wrap')
    
    # Compute curvature
    dx = np.gradient(x_smooth)
    dy = np.gradient(y_smooth)
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(dx * ddy - dy * ddx) / (dx**2 + dy**2 + 1e-10)**1.5
    curvature = (curvature - np.min(curvature)) / (np.max(curvature) - np.min(curvature) + 1e-10)
    
    # Find apex: min y
    apex_idx = np.argmin(y)
    apex_point = (int(x[apex_idx]), int(y[apex_idx]))
    
    # Find all peaks in curvature
    peaks, _ = find_peaks(curvature, prominence=0.0005)
    
    # Define basal region: y > max_y * (1 - delta_y_fraction)
    max_y = max(y)
    min_y = min(y)
    basal_threshold = max_y - delta_y_fraction * (max_y - min_y)
    basal_peaks = [i for i in peaks if y[i] >= basal_threshold]
    
    if len(basal_peaks) >= 2:
        # Sort peaks by curvature (highest to lowest)
        sorted_peaks = sorted(basal_peaks, key=lambda i: curvature[i], reverse=True)
        # Assign first highest peak as septal, second as lateral
        septal_idx = sorted_peaks[0]
        lateral_idx = sorted_peaks[1]
        
        # Verify they are on one line (similar y-coordinates)
        if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
            # Find a lateral peak with similar y-coordinate to septal
            septal_y = y[septal_idx]
            lateral_candidates = [i for i in basal_peaks if abs(y[i] - septal_y) <= y_tolerance]
            if len(lateral_candidates) >= 2:
                # Select the highest curvature peak among candidates, excluding septal
                lateral_candidates = [i for i in lateral_candidates if i != septal_idx]
                if lateral_candidates:
                    lateral_idx = max(lateral_candidates, key=lambda i: curvature[i])
                else:
                    raise ValueError("No suitable lateral peak found with similar y-coordinate")
            else:
                raise ValueError("Not enough peaks with similar y-coordinates")
    else:
        # Fallback: points with y > mean(y), min and max x
        mean_y = np.mean(y)
        lower_indices = [i for i in range(len(y)) if y[i] > mean_y]
        if len(lower_indices) >= 2:
            septal_idx = min(lower_indices, key=lambda i: x[i])
            lateral_idx = max(lower_indices, key=lambda i: x[i])
            # Verify y-alignment in fallback
            if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
                raise ValueError("Fallback septal and lateral points not on one line")
        else:
            raise ValueError("Not enough points for fallback")
    
    septal_point = (int(x[septal_idx]), int(y[septal_idx]))
    lateral_point = (int(x[lateral_idx]), int(y[lateral_idx]))
    
    # Mitral mid: midpoint between septal and lateral
    mitral_mid = ((septal_point[0] + lateral_point[0]) / 2, (septal_point[1] + lateral_point[1]) / 2)
    
    return apex_point, septal_point, lateral_point, mitral_mid

def compute_disk_diameters(seg_binary, apex, mitral_mid, pixel_spacing=(0.1, 0.1), nr_disks=20):
    """Compute disk diameters, long axis length, and dead space for a single view."""
    # Precompute filled LV mask
    contours, _ = cv2.findContours(seg_binary.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        return 0.0, [], [], 0.0
    
    max_contour = max(contours, key=cv2.contourArea)
    filled_mask = np.zeros_like(seg_binary)
    cv2.drawContours(filled_mask, [max_contour], -1, 1, thickness=cv2.FILLED)
    
    # Compute boundary
    kernel = np.ones((3,3), np.uint8)
    dilation = cv2.dilate(filled_mask, kernel)
    erosion = cv2.erode(filled_mask, kernel)
    boundary = (dilation - erosion > 0).astype(np.uint8)
    
    # Long axis vector
    dx = apex[0] - mitral_mid[0]
    dy = apex[1] - mitral_mid[1]
    L_px = np.sqrt(dx**2 + dy**2)
    L_cm = L_px * pixel_spacing[0]  # Assume isotropic spacing
    
    if L_cm == 0:
        return L_cm, [], [], 0.0
    
    # Perpendicular vector
    perp_vector = np.array([-dy, dx]) / L_px if L_px > 0 else np.array([1, 0])
    
    diameters_cm = []
    disk_info = []
    
    for i in range(nr_disks):
        t = (i + 0.5) / nr_disks
        center_x = mitral_mid[0] + t * dx
        center_y = mitral_mid[1] + t * dy
        
        left_point = None
        right_point = None
        
        # Search in both directions
        for direction in [-1, 1]:
            for dist in range(1, 150):
                px = int(round(center_x + direction * dist * perp_vector[0]))
                py = int(round(center_y + direction * dist * perp_vector[1]))
                
                if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                    break
                
                if boundary[py, px]:
                    if direction == -1:
                        left_point = (px, py)
                    else:
                        right_point = (px, py)
                    break
        
        if left_point and right_point:
            dx_px = right_point[0] - left_point[0]
            dy_px = right_point[1] - left_point[1]
            diameter_cm = np.sqrt((dx_px * pixel_spacing[0])**2 + (dy_px * pixel_spacing[1])**2)
            diameters_cm.append(diameter_cm)
            disk_info.append({
                'center': (center_x, center_y),
                'left': left_point,
                'right': right_point,
                'diameter_cm': diameter_cm
            })
        else:
            diameters_cm.append(0.0)
            disk_info.append({
                'center': (center_x, center_y),
                'left': None,
                'right': None,
                'diameter_cm': 0.0
            })
    
    # Calculate dead space area
    dead_area_cm2 = 0.0
    
    if len([d for d in disk_info if d['left'] and d['right']]) >= 2:
        # Create polygon from disk endpoints
        polygon_points = []
        
        # Add left points from base to apex
        for disk in disk_info:
            if disk['left']:
                polygon_points.append(disk['left'])
        
        # Add right points from apex to base
        for disk in reversed(disk_info):
            if disk['right']:
                polygon_points.append(disk['right'])
        
        # Create mask for disk-covered area
        disk_area_mask = np.zeros_like(filled_mask, dtype=np.uint8)
        if len(polygon_points) > 2:
            pts = np.array([polygon_points], dtype=np.int32)
            cv2.fillPoly(disk_area_mask, [pts], 1)
            
            # Calculate dead space as LV area not covered by disks
            dead_space_mask = filled_mask - disk_area_mask
            dead_space_mask = np.clip(dead_space_mask, 0, 1)  # Ensure binary
            
            # Convert to cm
            dead_area_px = np.sum(dead_space_mask)
            dead_area_cm2 = dead_area_px * pixel_spacing[0] * pixel_spacing[1]
    
    return L_cm, diameters_cm, disk_info, dead_area_cm2

def compute_biplane_volume(L4_cm, L2_cm, diameters4_cm, diameters2_cm, dead_area4_cm2, dead_area2_cm2, nr_disks=20):
    """Compute biplane Simpson's volume with dead space correction."""
    # Use average long axis length
    L_avg_cm = (L4_cm + L2_cm) / 2
    h_cm = L_avg_cm / nr_disks
    
    # Simpson's volume from disks
    total_volume = 0.0
    disk_volumes = []
    
    for i in range(nr_disks):
        a_i = diameters4_cm[i] if i < len(diameters4_cm) else 0
        b_i = diameters2_cm[i] if i < len(diameters2_cm) else 0
        disk_volume = (np.pi / 4) * a_i * b_i * h_cm
        total_volume += disk_volume
        disk_volumes.append(disk_volume)
    
    # Calculate dead volume
    avg_dead_area_cm2 = (dead_area4_cm2 + dead_area2_cm2) / 2
    dead_volume = avg_dead_area_cm2 * h_cm
    
    # Add dead volume to total
    total_volume_with_dead = total_volume + dead_volume
    
    return total_volume_with_dead

def visualize_processing(seg_data, landmarks, disk_info, view, phase, pixel_spacing):
    """Visualize the processing steps for one view/phase"""
    apex, septal, lateral, mitral_mid = landmarks
    
    # Create figure
    plt.figure(figsize=(15, 5))
    
    # Original image with landmarks
    plt.subplot(131)
    plt.imshow(seg_data, cmap='gray')
    plt.scatter(apex[0], apex[1], c='red', s=80, marker='x', label='Apex')
    plt.scatter(septal[0], septal[1], c='blue', s=80, marker='x', label='Septal')
    plt.scatter(lateral[0], lateral[1], c='green', s=80, marker='x', label='Lateral')
    plt.scatter(mitral_mid[0], mitral_mid[1], c='magenta', s=100, marker='+', label='Mitral Mid')
    plt.plot([apex[0], mitral_mid[0]], [apex[1], mitral_mid[1]], 'y--', label='Long Axis')
    plt.plot([septal[0], lateral[0]], [septal[1], lateral[1]], 'c--', label='Mitral Line')
    plt.title(f'{view} {phase} - Landmarks')
    plt.legend()
    
    # Measurement disks
    plt.subplot(132)
    plt.imshow(seg_data, cmap='gray')
    for disk in disk_info:
        if disk['left'] and disk['right']:
            plt.plot([disk['left'][0], disk['right'][0]], [disk['left'][1], disk['right'][1]], 'r-', linewidth=2)
        plt.scatter(disk['center'][0], disk['center'][1], c='yellow', s=20)
    plt.title(f'{view} {phase} - Measurement Disks')
    
    # Volume contribution per disk
    plt.subplot(133)
    if disk_info and 'diameter_cm' in disk_info[0]:
        diameters = [d['diameter_cm'] for d in disk_info]
        plt.plot(diameters, 'bo-')
        plt.title(f'{view} {phase} - Disk Diameters')
        plt.xlabel('Disk Number')
        plt.ylabel('Diameter (cm)')
        plt.grid(True)
    
    plt.tight_layout()
    plt.show()

def process_patient(patient_dir, visualize=False):
    """Process one patient with improved volume calculation"""
    patient_id = os.path.basename(patient_dir)
    print(f"\nProcessing Patient: {patient_id}")
    
    volumes = {'2CH': {'ED': None, 'ES': None}, '4CH': {'ED': None, 'ES': None}}
    all_landmarks = {}
    all_disk_info = {}
    
    for file in os.listdir(patient_dir):
        if not (file.endswith('_gt.niigz') or file.endswith('_gt.nii.gz')):
            continue
            
        try:
            # Determine view and phase
            view = '4CH' if '4ch' in file.lower() else '2CH' if '2ch' in file.lower() else None
            phase = 'ED' if 'ed' in file.lower() else 'ES' if 'es' in file.lower() else None
            
            if not view or not phase:
                continue
                
            print(f"\nProcessing {view} {phase}...")
            
            # Handle compressed files
            filepath = os.path.join(patient_dir, file)
            temp_file = None
            
            if file.endswith('.niigz'):
                temp_file = handle_compressed_file(filepath)
                gt_img = nib.load(temp_file)
            else:
                gt_img = nib.load(filepath)
                
            gt_data = gt_img.get_fdata()
            pixel_spacing = gt_img.header.get_zooms()[:2]
            
            # Create mask and rotate
            mask = np.zeros_like(gt_data, dtype=np.uint8)
            mask[gt_data == 1] = 255
            rotated_mask = rotate_image_clockwise(mask)
            
            # Detect landmarks
            apex, septal, lateral, mitral_mid = detect_landmarks(rotated_mask)
            
            # Calculate disk diameters and dead space
            L_cm, diameters_cm, disk_info, dead_area_cm2 = compute_disk_diameters(
                rotated_mask, apex, mitral_mid, pixel_spacing
            )
            
            # Store results
            volumes[view][phase] = {
                'volume': None,  # Will be calculated after both views are processed
                'L_cm': L_cm,
                'diameters_cm': diameters_cm,
                'dead_area_cm2': dead_area_cm2,
                'status': 'success'
            }
            
            all_landmarks[(view, phase)] = (apex, septal, lateral, mitral_mid)
            all_disk_info[(view, phase)] = disk_info
            
            # Visualize if requested
            if visualize:
                visualize_processing(rotated_mask, (apex, septal, lateral, mitral_mid), 
                                  disk_info, view, phase, pixel_spacing)
            
        except Exception as e:
            print(f"Error processing {file}: {str(e)}")
            if view and phase:
                volumes[view][phase] = {
                    'volume': None,
                    'status': f'failed: {str(e)}'
                }
        finally:
            if temp_file and os.path.exists(temp_file):
                os.unlink(temp_file)
    
    # Calculate volumes and EF if we have all required data
    missing = []
    for view in ['2CH', '4CH']:
        for phase in ['ED', 'ES']:
            if not volumes[view][phase] or volumes[view][phase]['status'] != 'success':
                missing.append(f"{view} {phase}")
    
    if missing:
        print(f"\nIncomplete data for EF calculation. Missing: {', '.join(missing)}")
    else:
        # Calculate EDV and ESV using biplane Simpson's method
        try:
            # ED volumes
            ed_vol = compute_biplane_volume(
                volumes['4CH']['ED']['L_cm'],
                volumes['2CH']['ED']['L_cm'],
                volumes['4CH']['ED']['diameters_cm'],
                volumes['2CH']['ED']['diameters_cm'],
                volumes['4CH']['ED']['dead_area_cm2'],
                volumes['2CH']['ED']['dead_area_cm2']
            )
            
            # ES volumes
            es_vol = compute_biplane_volume(
                volumes['4CH']['ES']['L_cm'],
                volumes['2CH']['ES']['L_cm'],
                volumes['4CH']['ES']['diameters_cm'],
                volumes['2CH']['ES']['diameters_cm'],
                volumes['4CH']['ES']['dead_area_cm2'],
                volumes['2CH']['ES']['dead_area_cm2']
            )
            
            # Calculate EF
            ef = ((ed_vol - es_vol) / ed_vol) * 100 if ed_vol > 0 else 0
            
            # Update volumes dictionary
            volumes['EDV'] = ed_vol
            volumes['ESV'] = es_vol
            volumes['EF'] = ef
            
            print(f"\nEF Calculation Successful:")
            print(f"EDV: {ed_vol:.2f} ml")
            print(f"ESV: {es_vol:.2f} ml")
            print(f"EF: {ef:.2f}%")
            
            # Final visualization
            if visualize:
                plt.figure(figsize=(10, 5))
                bars = plt.bar(['EDV', 'ESV'], [ed_vol, es_vol], color=['blue', 'red'])
                plt.ylabel('Volume (ml)')
                plt.title(f'Ejection Fraction Calculation\nEF = {ef:.1f}%')
                for bar in bars:
                    height = bar.get_height()
                    plt.text(bar.get_x() + bar.get_width()/2., height,
                            f'{height:.1f} ml', ha='center', va='bottom')
                plt.show()
                
        except Exception as e:
            print(f"Error calculating volumes: {str(e)}")
    
    return volumes, all_landmarks, all_disk_info

# Process one patient
base_dir = r"E:\echo heart\Resources"
patient_dirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

if patient_dirs:
    patient_volumes, landmarks, disk_info = process_patient(
        os.path.join(base_dir, patient_dirs[7]), 
        visualize=True
    )
else:
    print("No patient directories found")
```

```python
import os
import numpy as np
import nibabel as nib
import cv2
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
import math
import gzip
import shutil
import tempfile
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

def handle_compressed_file(filepath):
    """Handle both .nii.gz and .niigz files"""
    if filepath.endswith('.niigz'):
        with tempfile.NamedTemporaryFile(suffix='.nii.gz', delete=False) as tmp_file:
            with gzip.open(filepath, 'rb') as f_in:
                shutil.copyfileobj(f_in, tmp_file)
            return tmp_file.name
    return filepath

def rotate_image_clockwise(image, angle=90):
    """Rotate image clockwise with better boundary handling"""
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, -angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_NEAREST)
    return rotated

def detect_landmarks(seg_binary, delta_y_fraction=0.1, y_tolerance=10):
    """Improved landmark detection with curvature-based refinement."""
    # Convert to binary and find contours
    seg_binary = (seg_binary > 0).astype(np.uint8) * 255
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        raise ValueError("No LV contour found")
    
    # Use the largest contour
    contour = max(contours, key=cv2.contourArea).squeeze()
    x = contour[:, 0].astype(float)
    y = contour[:, 1].astype(float)
    
    # Smooth contour
    x_smooth = gaussian_filter1d(x, sigma=5, mode='wrap')
    y_smooth = gaussian_filter1d(y, sigma=5, mode='wrap')
    
    # Compute curvature
    dx = np.gradient(x_smooth)
    dy = np.gradient(y_smooth)
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(dx * ddy - dy * ddx) / (dx**2 + dy**2 + 1e-10)**1.5
    curvature = (curvature - np.min(curvature)) / (np.max(curvature) - np.min(curvature) + 1e-10)
    
    # Find apex: min y
    apex_idx = np.argmin(y)
    apex_point = (int(x[apex_idx]), int(y[apex_idx]))
    
    # Find all peaks in curvature
    peaks, _ = find_peaks(curvature, prominence=0.0005)
    
    # Define basal region: y > max_y * (1 - delta_y_fraction)
    max_y = max(y)
    min_y = min(y)
    basal_threshold = max_y - delta_y_fraction * (max_y - min_y)
    basal_peaks = [i for i in peaks if y[i] >= basal_threshold]
    
    if len(basal_peaks) >= 2:
        # Sort peaks by curvature (highest to lowest)
        sorted_peaks = sorted(basal_peaks, key=lambda i: curvature[i], reverse=True)
        # Assign first highest peak as septal, second as lateral
        septal_idx = sorted_peaks[0]
        lateral_idx = sorted_peaks[1]
        
        # Verify they are on one line (similar y-coordinates)
        if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
            # Find a lateral peak with similar y-coordinate to septal
            septal_y = y[septal_idx]
            lateral_candidates = [i for i in basal_peaks if abs(y[i] - septal_y) <= y_tolerance]
            if len(lateral_candidates) >= 2:
                # Select the highest curvature peak among candidates, excluding septal
                lateral_candidates = [i for i in lateral_candidates if i != septal_idx]
                if lateral_candidates:
                    lateral_idx = max(lateral_candidates, key=lambda i: curvature[i])
                else:
                    raise ValueError("No suitable lateral peak found with similar y-coordinate")
            else:
                raise ValueError("Not enough peaks with similar y-coordinates")
    else:
        # Fallback: points with y > mean(y), min and max x
        mean_y = np.mean(y)
        lower_indices = [i for i in range(len(y)) if y[i] > mean_y]
        if len(lower_indices) >= 2:
            septal_idx = min(lower_indices, key=lambda i: x[i])
            lateral_idx = max(lower_indices, key=lambda i: x[i])
            # Verify y-alignment in fallback
            if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
                raise ValueError("Fallback septal and lateral points not on one line")
        else:
            raise ValueError("Not enough points for fallback")
    
    septal_point = (int(x[septal_idx]), int(y[septal_idx]))
    lateral_point = (int(x[lateral_idx]), int(y[lateral_idx]))
    
    # Mitral mid: midpoint between septal and lateral
    mitral_mid = ((septal_point[0] + lateral_point[0]) / 2, (septal_point[1] + lateral_point[1]) / 2)
    
    return apex_point, septal_point, lateral_point, mitral_mid

def compute_disk_diameters(seg_binary, apex, mitral_mid, pixel_spacing=(0.1, 0.1), nr_disks=20):
    """Compute disk diameters, long axis length, and dead space for a single view."""
    # Precompute filled LV mask
    contours, _ = cv2.findContours(seg_binary.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        return 0.0, [], [], 0.0
    
    max_contour = max(contours, key=cv2.contourArea)
    filled_mask = np.zeros_like(seg_binary)
    cv2.drawContours(filled_mask, [max_contour], -1, 1, thickness=cv2.FILLED)
    
    # Compute boundary
    kernel = np.ones((3,3), np.uint8)
    dilation = cv2.dilate(filled_mask, kernel)
    erosion = cv2.erode(filled_mask, kernel)
    boundary = (dilation - erosion > 0).astype(np.uint8)
    
    # Long axis vector
    dx = apex[0] - mitral_mid[0]
    dy = apex[1] - mitral_mid[1]
    L_px = np.sqrt(dx**2 + dy**2)
    L_cm = L_px * pixel_spacing[0]  # Assume isotropic spacing
    
    if L_cm == 0:
        return L_cm, [], [], 0.0
    
    # Perpendicular vector
    perp_vector = np.array([-dy, dx]) / L_px if L_px > 0 else np.array([1, 0])
    
    diameters_cm = []
    disk_info = []
    
    for i in range(nr_disks):
        t = (i + 0.5) / nr_disks
        center_x = mitral_mid[0] + t * dx
        center_y = mitral_mid[1] + t * dy
        
        left_point = None
        right_point = None
        
        # Search in both directions
        for direction in [-1, 1]:
            for dist in range(1, 150):
                px = int(round(center_x + direction * dist * perp_vector[0]))
                py = int(round(center_y + direction * dist * perp_vector[1]))
                
                if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                    break
                
                if boundary[py, px]:
                    if direction == -1:
                        left_point = (px, py)
                    else:
                        right_point = (px, py)
                    break
        
        if left_point and right_point:
            dx_px = right_point[0] - left_point[0]
            dy_px = right_point[1] - left_point[1]
            diameter_cm = np.sqrt((dx_px * pixel_spacing[0])**2 + (dy_px * pixel_spacing[1])**2)
            diameters_cm.append(diameter_cm)
            disk_info.append({
                'center': (center_x, center_y),
                'left': left_point,
                'right': right_point,
                'diameter_cm': diameter_cm
            })
        else:
            diameters_cm.append(0.0)
            disk_info.append({
                'center': (center_x, center_y),
                'left': None,
                'right': None,
                'diameter_cm': 0.0
            })
    
    # Calculate dead space area
    dead_area_cm2 = 0.0
    
    if len([d for d in disk_info if d['left'] and d['right']]) >= 2:
        # Create polygon from disk endpoints
        polygon_points = []
        
        # Add left points from base to apex
        for disk in disk_info:
            if disk['left']:
                polygon_points.append(disk['left'])
        
        # Add right points from apex to base
        for disk in reversed(disk_info):
            if disk['right']:
                polygon_points.append(disk['right'])
        
        # Create mask for disk-covered area
        disk_area_mask = np.zeros_like(filled_mask, dtype=np.uint8)
        if len(polygon_points) > 2:
            pts = np.array([polygon_points], dtype=np.int32)
            cv2.fillPoly(disk_area_mask, [pts], 1)
            
            # Calculate dead space as LV area not covered by disks
            dead_space_mask = filled_mask - disk_area_mask
            dead_space_mask = np.clip(dead_space_mask, 0, 1)  # Ensure binary
            
            # Convert to cm
            dead_area_px = np.sum(dead_space_mask)
            dead_area_cm2 = dead_area_px * pixel_spacing[0] * pixel_spacing[1]
    
    return L_cm, diameters_cm, disk_info, dead_area_cm2

def compute_biplane_volume(L4_cm, L2_cm, diameters4_cm, diameters2_cm, dead_area4_cm2, dead_area2_cm2, nr_disks=20):
    """Compute biplane Simpson's volume with dead space correction."""
    # Use average long axis length
    L_avg_cm = (L4_cm + L2_cm) / 2
    h_cm = L_avg_cm / nr_disks
    
    # Simpson's volume from disks
    total_volume = 0.0
    disk_volumes = []
    
    for i in range(nr_disks):
        a_i = diameters4_cm[i] if i < len(diameters4_cm) else 0
        b_i = diameters2_cm[i] if i < len(diameters2_cm) else 0
        disk_volume = (np.pi / 4) * a_i * b_i * h_cm
        total_volume += disk_volume
        disk_volumes.append(disk_volume)
    
    # Calculate dead volume
    avg_dead_area_cm2 = (dead_area4_cm2 + dead_area2_cm2) / 2
    dead_volume = avg_dead_area_cm2 * h_cm
    
    # Add dead volume to total
    total_volume_with_dead = total_volume + dead_volume
    
    return total_volume_with_dead

def visualize_processing(seg_data, landmarks, disk_info, view, phase, pixel_spacing):
    """Visualize the processing steps for one view/phase"""
    apex, septal, lateral, mitral_mid = landmarks
    
    # Create figure
    plt.figure(figsize=(15, 5))
    
    # Original image with landmarks
    plt.subplot(131)
    plt.imshow(seg_data, cmap='gray')
    plt.scatter(apex[0], apex[1], c='red', s=80, marker='x', label='Apex')
    plt.scatter(septal[0], septal[1], c='blue', s=80, marker='x', label='Septal')
    plt.scatter(lateral[0], lateral[1], c='green', s=80, marker='x', label='Lateral')
    plt.scatter(mitral_mid[0], mitral_mid[1], c='magenta', s=100, marker='+', label='Mitral Mid')
    plt.plot([apex[0], mitral_mid[0]], [apex[1], mitral_mid[1]], 'y--', label='Long Axis')
    plt.plot([septal[0], lateral[0]], [septal[1], lateral[1]], 'c--', label='Mitral Line')
    plt.title(f'{view} {phase} - Landmarks')
    plt.legend()
    
    # Measurement disks
    plt.subplot(132)
    plt.imshow(seg_data, cmap='gray')
    for disk in disk_info:
        if disk['left'] and disk['right']:
            plt.plot([disk['left'][0], disk['right'][0]], [disk['left'][1], disk['right'][1]], 'r-', linewidth=2)
        plt.scatter(disk['center'][0], disk['center'][1], c='yellow', s=20)
    plt.title(f'{view} {phase} - Measurement Disks')
    
    # Volume contribution per disk
    plt.subplot(133)
    if disk_info and 'diameter_cm' in disk_info[0]:
        diameters = [d['diameter_cm'] for d in disk_info]
        plt.plot(diameters, 'bo-')
        plt.title(f'{view} {phase} - Disk Diameters')
        plt.xlabel('Disk Number')
        plt.ylabel('Diameter (cm)')
        plt.grid(True)
    
    plt.tight_layout()
    plt.show()

def process_patient(patient_dir, visualize=False):
    """Process one patient with improved volume calculation"""
    patient_id = os.path.basename(patient_dir)
    print(f"\nProcessing Patient: {patient_id}")
    
    volumes = {'2CH': {'ED': None, 'ES': None}, '4CH': {'ED': None, 'ES': None}}
    all_landmarks = {}
    all_disk_info = {}
    
    for file in os.listdir(patient_dir):
        if not (file.endswith('_gt.niigz') or file.endswith('_gt.nii.gz')):
            continue
            
        try:
            # Determine view and phase
            view = '4CH' if '4ch' in file.lower() else '2CH' if '2ch' in file.lower() else None
            phase = 'ED' if 'ed' in file.lower() else 'ES' if 'es' in file.lower() else None
            
            if not view or not phase:
                continue
                
            print(f"\nProcessing {view} {phase}...")
            
            # Handle compressed files
            filepath = os.path.join(patient_dir, file)
            temp_file = None
            
            if file.endswith('.niigz'):
                temp_file = handle_compressed_file(filepath)
                gt_img = nib.load(temp_file)
            else:
                gt_img = nib.load(filepath)
                
            gt_data = gt_img.get_fdata()
            pixel_spacing = gt_img.header.get_zooms()[:2]
            
            # Create mask and rotate
            mask = np.zeros_like(gt_data, dtype=np.uint8)
            mask[gt_data == 1] = 255
            rotated_mask = rotate_image_clockwise(mask)
            
            # Detect landmarks
            apex, septal, lateral, mitral_mid = detect_landmarks(rotated_mask)
            
            # Calculate disk diameters and dead space
            L_cm, diameters_cm, disk_info, dead_area_cm2 = compute_disk_diameters(
                rotated_mask, apex, mitral_mid, pixel_spacing
            )
            
            # Store results
            volumes[view][phase] = {
                'volume': None,  # Will be calculated after both views are processed
                'L_cm': L_cm,
                'diameters_cm': diameters_cm,
                'dead_area_cm2': dead_area_cm2,
                'status': 'success'
            }
            
            all_landmarks[(view, phase)] = (apex, septal, lateral, mitral_mid)
            all_disk_info[(view, phase)] = disk_info
            
            # Visualize if requested
            if visualize:
                visualize_processing(rotated_mask, (apex, septal, lateral, mitral_mid), 
                                  disk_info, view, phase, pixel_spacing)
            
        except Exception as e:
            print(f"Error processing {file}: {str(e)}")
            if view and phase:
                volumes[view][phase] = {
                    'volume': None,
                    'status': f'failed: {str(e)}'
                }
        finally:
            if temp_file and os.path.exists(temp_file):
                os.unlink(temp_file)
    
    # Calculate volumes and EF if we have all required data
    missing = []
    for view in ['2CH', '4CH']:
        for phase in ['ED', 'ES']:
            if not volumes[view][phase] or volumes[view][phase]['status'] != 'success':
                missing.append(f"{view} {phase}")
    
    if missing:
        print(f"\nIncomplete data for EF calculation. Missing: {', '.join(missing)}")
    else:
        # Calculate EDV and ESV using biplane Simpson's method
        try:
            # ED volumes
            ed_vol = compute_biplane_volume(
                volumes['4CH']['ED']['L_cm'],
                volumes['2CH']['ED']['L_cm'],
                volumes['4CH']['ED']['diameters_cm'],
                volumes['2CH']['ED']['diameters_cm'],
                volumes['4CH']['ED']['dead_area_cm2'],
                volumes['2CH']['ED']['dead_area_cm2']
            )
            
            # ES volumes
            es_vol = compute_biplane_volume(
                volumes['4CH']['ES']['L_cm'],
                volumes['2CH']['ES']['L_cm'],
                volumes['4CH']['ES']['diameters_cm'],
                volumes['2CH']['ES']['diameters_cm'],
                volumes['4CH']['ES']['dead_area_cm2'],
                volumes['2CH']['ES']['dead_area_cm2']
            )
            
            # Calculate EF
            ef = ((ed_vol - es_vol) / ed_vol) * 100 if ed_vol > 0 else 0
            
            # Update volumes dictionary
            volumes['EDV'] = ed_vol
            volumes['ESV'] = es_vol
            volumes['EF'] = ef
            
            print(f"\nEF Calculation Successful:")
            print(f"EDV: {ed_vol:.2f} ml")
            print(f"ESV: {es_vol:.2f} ml")
            print(f"EF: {ef:.2f}%")
            
            # Final visualization
            if visualize:
                plt.figure(figsize=(10, 5))
                bars = plt.bar(['EDV', 'ESV'], [ed_vol, es_vol], color=['blue', 'red'])
                plt.ylabel('Volume (ml)')
                plt.title(f'Ejection Fraction Calculation\nEF = {ef:.1f}%')
                for bar in bars:
                    height = bar.get_height()
                    plt.text(bar.get_x() + bar.get_width()/2., height,
                            f'{height:.1f} ml', ha='center', va='bottom')
                plt.show()
                
        except Exception as e:
            print(f"Error calculating volumes: {str(e)}")
    
    return volumes, all_landmarks, all_disk_info

# Process one patient
base_dir = r"E:\echo heart\Resources"
patient_dirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

if patient_dirs:
    patient_volumes, landmarks, disk_info = process_patient(
        os.path.join(base_dir, patient_dirs[5]), 
        visualize=True
    )
else:
    print("No patient directories found")
```

```python
import os
import numpy as np
import nibabel as nib
import cv2
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from scipy.ndimage import gaussian_filter1d
import math
import gzip
import shutil
import tempfile
import matplotlib as mpl
from matplotlib.colors import LinearSegmentedColormap

def handle_compressed_file(filepath):
    """Handle both .nii.gz and .niigz files"""
    if filepath.endswith('.niigz'):
        with tempfile.NamedTemporaryFile(suffix='.nii.gz', delete=False) as tmp_file:
            with gzip.open(filepath, 'rb') as f_in:
                shutil.copyfileobj(f_in, tmp_file)
            return tmp_file.name
    return filepath

def rotate_image_clockwise(image, angle=90):
    """Rotate image clockwise with better boundary handling"""
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, -angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h), flags=cv2.INTER_NEAREST)
    return rotated

def detect_landmarks(seg_binary, delta_y_fraction=0.1, y_tolerance=10):
    """Improved landmark detection with curvature-based refinement."""
    # Convert to binary and find contours
    seg_binary = (seg_binary > 0).astype(np.uint8) * 255
    contours, _ = cv2.findContours(seg_binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        raise ValueError("No LV contour found")
    
    # Use the largest contour
    contour = max(contours, key=cv2.contourArea).squeeze()
    x = contour[:, 0].astype(float)
    y = contour[:, 1].astype(float)
    
    # Smooth contour
    x_smooth = gaussian_filter1d(x, sigma=5, mode='wrap')
    y_smooth = gaussian_filter1d(y, sigma=5, mode='wrap')
    
    # Compute curvature
    dx = np.gradient(x_smooth)
    dy = np.gradient(y_smooth)
    ddx = np.gradient(dx)
    ddy = np.gradient(dy)
    curvature = np.abs(dx * ddy - dy * ddx) / (dx**2 + dy**2 + 1e-10)**1.5
    curvature = (curvature - np.min(curvature)) / (np.max(curvature) - np.min(curvature) + 1e-10)
    
    # Find apex: min y
    apex_idx = np.argmin(y)
    apex_point = (int(x[apex_idx]), int(y[apex_idx]))
    
    # Find all peaks in curvature
    peaks, _ = find_peaks(curvature, prominence=0.0005)
    
    # Define basal region: y > max_y * (1 - delta_y_fraction)
    max_y = max(y)
    min_y = min(y)
    basal_threshold = max_y - delta_y_fraction * (max_y - min_y)
    basal_peaks = [i for i in peaks if y[i] >= basal_threshold]
    
    if len(basal_peaks) >= 2:
        # Sort peaks by curvature (highest to lowest)
        sorted_peaks = sorted(basal_peaks, key=lambda i: curvature[i], reverse=True)
        # Assign first highest peak as septal, second as lateral
        septal_idx = sorted_peaks[0]
        lateral_idx = sorted_peaks[1]
        
        # Verify they are on one line (similar y-coordinates)
        if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
            # Find a lateral peak with similar y-coordinate to septal
            septal_y = y[septal_idx]
            lateral_candidates = [i for i in basal_peaks if abs(y[i] - septal_y) <= y_tolerance]
            if len(lateral_candidates) >= 2:
                # Select the highest curvature peak among candidates, excluding septal
                lateral_candidates = [i for i in lateral_candidates if i != septal_idx]
                if lateral_candidates:
                    lateral_idx = max(lateral_candidates, key=lambda i: curvature[i])
                else:
                    raise ValueError("No suitable lateral peak found with similar y-coordinate")
            else:
                raise ValueError("Not enough peaks with similar y-coordinates")
    else:
        # Fallback: points with y > mean(y), min and max x
        mean_y = np.mean(y)
        lower_indices = [i for i in range(len(y)) if y[i] > mean_y]
        if len(lower_indices) >= 2:
            septal_idx = min(lower_indices, key=lambda i: x[i])
            lateral_idx = max(lower_indices, key=lambda i: x[i])
            # Verify y-alignment in fallback
            if abs(y[septal_idx] - y[lateral_idx]) > y_tolerance:
                raise ValueError("Fallback septal and lateral points not on one line")
        else:
            raise ValueError("Not enough points for fallback")
    
    septal_point = (int(x[septal_idx]), int(y[septal_idx]))
    lateral_point = (int(x[lateral_idx]), int(y[lateral_idx]))
    
    # Mitral mid: midpoint between septal and lateral
    mitral_mid = ((septal_point[0] + lateral_point[0]) / 2, (septal_point[1] + lateral_point[1]) / 2)
    
    return apex_point, septal_point, lateral_point, mitral_mid

def compute_disk_diameters(seg_binary, apex, mitral_mid, pixel_spacing=(0.1, 0.1), nr_disks=20):
    """Compute disk diameters, long axis length, and dead space for a single view."""
    # Precompute filled LV mask
    contours, _ = cv2.findContours(seg_binary.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if not contours:
        return 0.0, [], [], 0.0
    
    max_contour = max(contours, key=cv2.contourArea)
    filled_mask = np.zeros_like(seg_binary)
    cv2.drawContours(filled_mask, [max_contour], -1, 1, thickness=cv2.FILLED)
    
    # Compute boundary
    kernel = np.ones((3,3), np.uint8)
    dilation = cv2.dilate(filled_mask, kernel)
    erosion = cv2.erode(filled_mask, kernel)
    boundary = (dilation - erosion > 0).astype(np.uint8)
    
    # Long axis vector
    dx = apex[0] - mitral_mid[0]
    dy = apex[1] - mitral_mid[1]
    L_px = np.sqrt(dx**2 + dy**2)
    L_cm = L_px * pixel_spacing[0]  # Assume isotropic spacing
    
    if L_cm == 0:
        return L_cm, [], [], 0.0
    
    # Perpendicular vector
    perp_vector = np.array([-dy, dx]) / L_px if L_px > 0 else np.array([1, 0])
    
    diameters_cm = []
    disk_info = []
    
    for i in range(nr_disks):
        t = (i + 0.5) / nr_disks
        center_x = mitral_mid[0] + t * dx
        center_y = mitral_mid[1] + t * dy
        
        left_point = None
        right_point = None
        
        # Search in both directions
        for direction in [-1, 1]:
            for dist in range(1, 150):
                px = int(round(center_x + direction * dist * perp_vector[0]))
                py = int(round(center_y + direction * dist * perp_vector[1]))
                
                if px < 0 or px >= seg_binary.shape[1] or py < 0 or py >= seg_binary.shape[0]:
                    break
                
                if boundary[py, px]:
                    if direction == -1:
                        left_point = (px, py)
                    else:
                        right_point = (px, py)
                    break
        
        if left_point and right_point:
            dx_px = right_point[0] - left_point[0]
            dy_px = right_point[1] - left_point[1]
            diameter_cm = np.sqrt((dx_px * pixel_spacing[0])**2 + (dy_px * pixel_spacing[1])**2)
            diameters_cm.append(diameter_cm)
            disk_info.append({
                'center': (center_x, center_y),
                'left': left_point,
                'right': right_point,
                'diameter_cm': diameter_cm
            })
        else:
            diameters_cm.append(0.0)
            disk_info.append({
                'center': (center_x, center_y),
                'left': None,
                'right': None,
                'diameter_cm': 0.0
            })
    
    # Calculate dead space area
    dead_area_cm2 = 0.0
    
    if len([d for d in disk_info if d['left'] and d['right']]) >= 2:
        # Create polygon from disk endpoints
        polygon_points = []
        
        # Add left points from base to apex
        for disk in disk_info:
            if disk['left']:
                polygon_points.append(disk['left'])
        
        # Add right points from apex to base
        for disk in reversed(disk_info):
            if disk['right']:
                polygon_points.append(disk['right'])
        
        # Create mask for disk-covered area
        disk_area_mask = np.zeros_like(filled_mask, dtype=np.uint8)
        if len(polygon_points) > 2:
            pts = np.array([polygon_points], dtype=np.int32)
            cv2.fillPoly(disk_area_mask, [pts], 1)
            
            # Calculate dead space as LV area not covered by disks
            dead_space_mask = filled_mask - disk_area_mask
            dead_space_mask = np.clip(dead_space_mask, 0, 1)  # Ensure binary
            
            # Convert to cm
            dead_area_px = np.sum(dead_space_mask)
            dead_area_cm2 = dead_area_px * pixel_spacing[0] * pixel_spacing[1]
    
    return L_cm, diameters_cm, disk_info, dead_area_cm2

def compute_biplane_volume(L4_cm, L2_cm, diameters4_cm, diameters2_cm, dead_area4_cm2, dead_area2_cm2, nr_disks=20):
    """Compute biplane Simpson's volume with dead space correction."""
    # Use average long axis length
    L_avg_cm = (L4_cm + L2_cm) / 2
    h_cm = L_avg_cm / nr_disks
    
    # Simpson's volume from disks
    total_volume = 0.0
    disk_volumes = []
    
    for i in range(nr_disks):
        a_i = diameters4_cm[i] if i < len(diameters4_cm) else 0
        b_i = diameters2_cm[i] if i < len(diameters2_cm) else 0
        disk_volume = (np.pi / 4) * a_i * b_i * h_cm
        total_volume += disk_volume
        disk_volumes.append(disk_volume)
    
    # Calculate dead volume
    avg_dead_area_cm2 = (dead_area4_cm2 + dead_area2_cm2) / 2
    dead_volume = avg_dead_area_cm2 * h_cm
    
    # Add dead volume to total
    total_volume_with_dead = total_volume + dead_volume
    
    return total_volume_with_dead

def visualize_processing(seg_data, landmarks, disk_info, view, phase, pixel_spacing):
    """Visualize the processing steps for one view/phase"""
    apex, septal, lateral, mitral_mid = landmarks
    
    # Create figure
    plt.figure(figsize=(15, 5))
    
    # Original image with landmarks
    plt.subplot(131)
    plt.imshow(seg_data, cmap='gray')
    plt.scatter(apex[0], apex[1], c='red', s=80, marker='x', label='Apex')
    plt.scatter(septal[0], septal[1], c='blue', s=80, marker='x', label='Septal')
    plt.scatter(lateral[0], lateral[1], c='green', s=80, marker='x', label='Lateral')
    plt.scatter(mitral_mid[0], mitral_mid[1], c='magenta', s=100, marker='+', label='Mitral Mid')
    plt.plot([apex[0], mitral_mid[0]], [apex[1], mitral_mid[1]], 'y--', label='Long Axis')
    plt.plot([septal[0], lateral[0]], [septal[1], lateral[1]], 'c--', label='Mitral Line')
    plt.title(f'{view} {phase} - Landmarks')
    plt.legend()
    
    # Measurement disks
    plt.subplot(132)
    plt.imshow(seg_data, cmap='gray')
    for disk in disk_info:
        if disk['left'] and disk['right']:
            plt.plot([disk['left'][0], disk['right'][0]], [disk['left'][1], disk['right'][1]], 'r-', linewidth=2)
        plt.scatter(disk['center'][0], disk['center'][1], c='yellow', s=20)
    plt.title(f'{view} {phase} - Measurement Disks')
    
    # Volume contribution per disk
    plt.subplot(133)
    if disk_info and 'diameter_cm' in disk_info[0]:
        diameters = [d['diameter_cm'] for d in disk_info]
        plt.plot(diameters, 'bo-')
        plt.title(f'{view} {phase} - Disk Diameters')
        plt.xlabel('Disk Number')
        plt.ylabel('Diameter (cm)')
        plt.grid(True)
    
    plt.tight_layout()
    plt.show()

def process_patient(patient_dir, visualize=False):
    """Process one patient with improved volume calculation"""
    patient_id = os.path.basename(patient_dir)
    print(f"\nProcessing Patient: {patient_id}")
    
    volumes = {'2CH': {'ED': None, 'ES': None}, '4CH': {'ED': None, 'ES': None}}
    all_landmarks = {}
    all_disk_info = {}
    
    for file in os.listdir(patient_dir):
        if not (file.endswith('_gt.niigz') or file.endswith('_gt.nii.gz')):
            continue
            
        try:
            # Determine view and phase
            view = '4CH' if '4ch' in file.lower() else '2CH' if '2ch' in file.lower() else None
            phase = 'ED' if 'ed' in file.lower() else 'ES' if 'es' in file.lower() else None
            
            if not view or not phase:
                continue
                
            print(f"\nProcessing {view} {phase}...")
            
            # Handle compressed files
            filepath = os.path.join(patient_dir, file)
            temp_file = None
            
            if file.endswith('.niigz'):
                temp_file = handle_compressed_file(filepath)
                gt_img = nib.load(temp_file)
            else:
                gt_img = nib.load(filepath)
                
            gt_data = gt_img.get_fdata()
            pixel_spacing = gt_img.header.get_zooms()[:2]
            
            # Create mask and rotate
            mask = np.zeros_like(gt_data, dtype=np.uint8)
            mask[gt_data == 1] = 255
            rotated_mask = rotate_image_clockwise(mask)
            
            # Detect landmarks
            apex, septal, lateral, mitral_mid = detect_landmarks(rotated_mask)
            
            # Calculate disk diameters and dead space
            L_cm, diameters_cm, disk_info, dead_area_cm2 = compute_disk_diameters(
                rotated_mask, apex, mitral_mid, pixel_spacing
            )
            
            # Store results
            volumes[view][phase] = {
                'volume': None,  # Will be calculated after both views are processed
                'L_cm': L_cm,
                'diameters_cm': diameters_cm,
                'dead_area_cm2': dead_area_cm2,
                'status': 'success'
            }
            
            all_landmarks[(view, phase)] = (apex, septal, lateral, mitral_mid)
            all_disk_info[(view, phase)] = disk_info
            
            # Visualize if requested
            if visualize:
                visualize_processing(rotated_mask, (apex, septal, lateral, mitral_mid), 
                                  disk_info, view, phase, pixel_spacing)
            
        except Exception as e:
            print(f"Error processing {file}: {str(e)}")
            if view and phase:
                volumes[view][phase] = {
                    'volume': None,
                    'status': f'failed: {str(e)}'
                }
        finally:
            if temp_file and os.path.exists(temp_file):
                os.unlink(temp_file)
    
    # Calculate volumes and EF if we have all required data
    missing = []
    for view in ['2CH', '4CH']:
        for phase in ['ED', 'ES']:
            if not volumes[view][phase] or volumes[view][phase]['status'] != 'success':
                missing.append(f"{view} {phase}")
    
    if missing:
        print(f"\nIncomplete data for EF calculation. Missing: {', '.join(missing)}")
    else:
        # Calculate EDV and ESV using biplane Simpson's method
        try:
            # ED volumes
            ed_vol = compute_biplane_volume(
                volumes['4CH']['ED']['L_cm'],
                volumes['2CH']['ED']['L_cm'],
                volumes['4CH']['ED']['diameters_cm'],
                volumes['2CH']['ED']['diameters_cm'],
                volumes['4CH']['ED']['dead_area_cm2'],
                volumes['2CH']['ED']['dead_area_cm2']
            )
            
            # ES volumes
            es_vol = compute_biplane_volume(
                volumes['4CH']['ES']['L_cm'],
                volumes['2CH']['ES']['L_cm'],
                volumes['4CH']['ES']['diameters_cm'],
                volumes['2CH']['ES']['diameters_cm'],
                volumes['4CH']['ES']['dead_area_cm2'],
                volumes['2CH']['ES']['dead_area_cm2']
            )
            
            # Calculate EF
            ef = ((ed_vol - es_vol) / ed_vol) * 100 if ed_vol > 0 else 0
            
            # Update volumes dictionary
            volumes['EDV'] = ed_vol
            volumes['ESV'] = es_vol
            volumes['EF'] = ef
            
            print(f"\nEF Calculation Successful:")
            print(f"EDV: {ed_vol:.2f} ml")
            print(f"ESV: {es_vol:.2f} ml")
            print(f"EF: {ef:.2f}%")
            
            # Final visualization
            if visualize:
                plt.figure(figsize=(10, 5))
                bars = plt.bar(['EDV', 'ESV'], [ed_vol, es_vol], color=['blue', 'red'])
                plt.ylabel('Volume (ml)')
                plt.title(f'Ejection Fraction Calculation\nEF = {ef:.1f}%')
                for bar in bars:
                    height = bar.get_height()
                    plt.text(bar.get_x() + bar.get_width()/2., height,
                            f'{height:.1f} ml', ha='center', va='bottom')
                plt.show()
                
        except Exception as e:
            print(f"Error calculating volumes: {str(e)}")
    
    return volumes, all_landmarks, all_disk_info

# Process one patient
base_dir = r"E:\echo heart\Resources"
patient_dirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]

if patient_dirs:
    patient_volumes, landmarks, disk_info = process_patient(
        os.path.join(base_dir, patient_dirs[4]), 
        visualize=True
    )
else:
    print("No patient directories found")
```
